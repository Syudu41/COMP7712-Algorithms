## Question 1 (Non-coding) --- Program Correctness

Consider the following recursive function:

```python
def count_occurrences(L, x):
    if L == []:
        return 0
    if L[0] == x:
        return 1 + count_occurrences(L[1:], x)
    else:
        return count_occurrences(L[1:], x)
```

Use mathematical induction to argue that this function is correct. Specifically:

**(a)** State what it means for this function to be correct (i.e., what does this function compute?).

**(b)** Prove the **base case**: Explain why the function is correct for the smallest input.

**(c)** Prove the **inductive step**: Assume the recursive calls on line 5 and line 7 return correct answers for lists with fewer than $n$ items. Explain why the function then returns the correct answer for a list with $n$ items.

**Your Answer:**

(a) The function `count_occurrences(L, x)` returns the number of times the value `x` appears in the list `L`. The function is correct if, for every list `L` and value `x`, it returns the count of elements in `L` that are equal to `x`.

(b) **Base case:** The smallest input is the empty list `L = []`. An empty list contains no elements, so the number of occurrences of any value `x` in it is 0. The function returns 0 when `L == []` (line 3). Therefore, the function is correct for the base case.

(c) **Inductive step:** Assume `count_occurrences` returns the correct answer for any list with fewer than $n$ items (this is our inductive hypothesis). Now consider a list `L` with $n$ items ($n \ge 1$). The sublist `L[1:]` has $n - 1$ items, which is fewer than $n$, so by the inductive hypothesis the recursive call `count_occurrences(L[1:], x)` correctly returns the number of occurrences of `x` in `L[1:]`.

There are two cases:
- If `L[0] == x`: The total occurrences of `x` in `L` is 1 (for `L[0]`) plus the occurrences of `x` in the rest of the list `L[1:]`. The function returns `1 + count_occurrences(L[1:], x)`, which by the inductive hypothesis equals `1 +` (correct count in `L[1:]`) = correct total count. $\checkmark$
- If `L[0] != x`: The first element does not contribute to the count, so the total occurrences of `x` in `L` equals the occurrences of `x` in `L[1:]`. The function returns `count_occurrences(L[1:], x)`, which by the inductive hypothesis is the correct count. $\checkmark$

In both cases the function returns the correct answer for a list of size $n$. By the principle of mathematical induction, the function is correct for all lists.

## Question 2 (Non-coding) --- Program Correctness

Consider the following recursive function that computes the maximum element in a list by dividing it in half:

```python
def find_max(L):
    if len(L) == 1:
        return L[0]
    mid = len(L) // 2
    left_max = find_max(L[:mid])
    right_max = find_max(L[mid:])
    if left_max > right_max:
        return left_max
    else:
        return right_max
```

**(a)** Does this function use basic induction or strong induction for its correctness argument? Explain why.

**(b)** Prove the **base case**: Explain why the function is correct for the smallest input.

**(c)** Prove the **inductive step**: Assume `find_max` returns the correct answer for all lists with fewer than $n$ items. Explain why it returns the correct answer for a list with $n$ items ($n \ge 2$).

**Your Answer:**

(a) This function requires **strong induction**. In basic induction, the inductive hypothesis only assumes correctness for lists of size $n - 1$. However, `find_max` splits the list at the midpoint, producing two sublists of size $\lfloor n/2 \rfloor$ and $\lceil n/2 \rceil$. These sublists can be *any* size smaller than $n$ (not just $n - 1$). Therefore, we need the strong inductive hypothesis: assume `find_max` is correct for *all* lists with fewer than $n$ elements.

(b) **Base case:** The smallest valid input is a list with exactly 1 element. When `len(L) == 1`, the function returns `L[0]`. The maximum of a single-element list is that element itself. Therefore, the function is correct for the base case.

(c) **Inductive step:** Assume `find_max` returns the correct maximum for all lists with fewer than $n$ elements (strong inductive hypothesis). Consider a list `L` with $n$ items where $n \ge 2$.

The function computes `mid = len(L) // 2`. Since $n \ge 2$, we have $1 \le mid < n$, so `L[:mid]` has `mid` elements and `L[mid:]` has `n - mid` elements. Both sublists are non-empty and both have strictly fewer than $n$ elements.

By the strong inductive hypothesis:
- `left_max = find_max(L[:mid])` correctly returns the maximum of the left half.
- `right_max = find_max(L[mid:])` correctly returns the maximum of the right half.

Every element in `L` is either in the left half or the right half (the two halves partition `L`). Therefore, the overall maximum of `L` is the larger of `left_max` and `right_max`. The if/else on lines 7--9 returns exactly that larger value.

Therefore, `find_max(L)` returns the correct maximum for a list of size $n$. By strong induction, the function is correct for all non-empty lists.

Therefore, `L[0] * product(L[1:])` = $L[0] \times L[1] \times \cdots \times L[n-1]$, which is the product of all elements in `L`. $\checkmark$

## Question 4 (Non-coding) --- Running Time

For each of the following code fragments, determine the running time equation $T(n)$ using constant-step notation ($a$, $b$, etc.).

**(a)**
```python
total = 0
for i in range(n):
    total += i
```

**(b)**
```python
total = 0
for i in range(n):
    for j in range(n):
        total += i + j
```

**(c)**
```python

    for j in range(i):
        total += i * j
```

For part (c), give an upper bound and a lower bound for $T(n)$.

**Your Answer:**

**(a)** $T(n) = a + bn$

- `total = 0` takes $a$ steps (one assignment).
- The loop runs $n$ times, and each iteration does a constant amount of work ($b$ steps for the addition and assignment).
- Total: $T(n) = a + bn$.

**(b)** $T(n) = a + bn^2$

- `total = 0` takes $a$ steps.
- The outer loop runs $n$ times. For each iteration of the outer loop, the inner loop runs $n$ times, doing $b$ steps each time.
- The inner loop body executes $n \times n = n^2$ times total.
- Total: $T(n) = a + bn^2$.

**(c)** Lower bound and upper bound:

The inner loop runs `i` times for each value of `i`. The total number of inner loop body executions is:
$$0 + 1 + 2 + \cdots + (n-1) = \frac{n(n-1)}{2}$$

So $T(n) = a + b \cdot \frac{n(n-1)}{2}$.

- **Upper bound:** $\frac{n(n-1)}{2} < \frac{n^2}{2}$, so $T(n) < a + \frac{b}{2} n^2$. Thus $T(n)$ is bounded above by something proportional to $n^2$.
- **Lower bound:** $\frac{n(n-1)}{2} = \frac{n^2 - n}{2} \ge \frac{n^2}{4}$ for $n \ge 2$ (since $n \ge n/2$ implies $n - n/2 \ge n/2$, so $n(n-1) \ge n \cdot n/2 = n^2/2$). Thus $T(n) \ge a + \frac{b}{4} n^2$, so $T(n)$ is bounded below by something proportional to $n^2$.

## Question 5 (Non-coding) --- Running Time

Consider the following function:

```python
def mystery(L):
    n = len(L)
    count = 0
    for i in range(n):
        for j in range(i+1, n):
            if L[i] > L[j]:
                count += 1
    return count
```

**(a)** What does this function compute? (Describe its output in plain English.)

**(b)** The inner loop runs a variable number of times. Write an upper bound for $T(n)$.

**(c)** Calculate the exact total number of times the inner loop body executes across all iterations of the outer loop. (Hint: use the formula $0 + 1 + 2 + \cdots + (n-1) = \frac{n(n-1)}{2}$.)

**Your Answer:**

(a) This function counts the number of **inversions** in the list. An inversion is a pair of indices $(i, j)$ with $i < j$ but $L[i] > L[j]$ --- that is, a pair of elements that are "out of order." For a sorted list, the count is 0. For a reverse-sorted list, the count is maximized.

(b) For each value of `i`, the inner loop runs at most $n - 1$ times (when `i = 0`). Since the outer loop runs $n$ times, an upper bound is:
$$T(n) \le a + b \cdot n \cdot (n - 1) < a + bn^2$$
So $T(n)$ is at most proportional to $n^2$.

(c) When `i = 0`, the inner loop runs $n - 1$ times. When `i = 1`, it runs $n - 2$ times. In general, when `i = k`, it runs $n - 1 - k$ times. When `i = n - 1`, it runs 0 times.

The total number of inner loop body executions is:
$$(n-1) + (n-2) + \cdots + 1 + 0 = \sum_{k=0}^{n-1}(n - 1 - k) = 0 + 1 + 2 + \cdots + (n-1) = \frac{n(n-1)}{2}$$

So the inner loop body executes exactly $\frac{n(n-1)}{2}$ times.

## Question 10 (Non-coding) --- Putting It All Together

Consider the following function:

```python
def has_duplicates(L):
    n = len(L)
    for i in range(n):
        for j in range(i+1, n):
            if L[i] == L[j]:
                return True
    return False
```

**(a)** Explain what this function does.

**(b)** What is the **best case** running time (when does the function finish fastest)? Give $T(n)$ using constant-step notation.

**(c)** What is the **worst case** running time (when does the function take the longest)? Give $T(n)$ using constant-step notation.

**(d)** Show that the worst-case running time is in $\Theta(n^2)$.

**Your Answer:**

**(a)** This function checks whether the list `L` contains any duplicate elements. It compares every pair of distinct elements. If it finds any pair `(L[i], L[j])` with `i < j` where `L[i] == L[j]`, it returns `True` (duplicates exist). If no such pair is found after checking all pairs, it returns `False` (all elements are unique).

**(b) Best case:** The function finishes fastest when the first two elements are equal (i.e., `L[0] == L[1]`). In this case, the outer loop runs once (`i = 0`), the inner loop runs once (`j = 1`), the comparison is true, and the function immediately returns `True`.

$$T_{\text{best}}(n) = a$$

where $a$ is a constant (a fixed number of steps: computing `len(L)`, one iteration of each loop, one comparison, and the return). The best case is $O(1)$ --- constant time.

**(c) Worst case:** The function takes the longest when there are no duplicates (or the only duplicates are at the very end), so it must check every pair before returning `False`. The inner loop body executes:
$$(n-1) + (n-2) + \cdots + 1 + 0 = \frac{n(n-1)}{2} \text{ times}$$

$$T_{\text{worst}}(n) = a + b \cdot \frac{n(n-1)}{2}$$

where $a$ accounts for the setup steps and $b$ is the cost of each inner loop iteration.

**(d)** We show $T_{\text{worst}}(n) = a + b \cdot \frac{n(n-1)}{2} \in \Theta(n^2)$ by finding $c_1$ and $c_2$.

**Upper bound ($O$):** $\frac{n(n-1)}{2} = \frac{n^2 - n}{2} \le \frac{n^2}{2}$ for all $n \ge 1$. So:
$$T_{\text{worst}}(n) = a + b \cdot \frac{n(n-1)}{2} \le a + \frac{b}{2} n^2 \le (a + \frac{b}{2}) n^2$$
for $n \ge 1$ (since $a \le a \cdot n^2$). Choose $c_2 = a + \frac{b}{2}$.

**Lower bound ($\Omega$):** For $n \ge 2$: $\frac{n(n-1)}{2} \ge \frac{n \cdot n/2}{2} = \frac{n^2}{4}$. So:
$$T_{\text{worst}}(n) = a + b \cdot \frac{n(n-1)}{2} \ge \frac{b}{4} n^2$$
Choose $c_1 = \frac{b}{4}$.

Therefore $\frac{b}{4} n^2 \le T_{\text{worst}}(n) \le (a + \frac{b}{2}) n^2$ for all $n \ge 2$, confirming $T_{\text{worst}}(n) \in \Theta(n^2)$. $\checkmark$

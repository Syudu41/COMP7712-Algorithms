## Question 1 (Non-coding)

Consider the following recursive function that computes the sum of digits of a non-negative integer:

```python
def sum_digits(n):
    if n < 10:
        return n
    return n % 10 + sum_digits(n // 10)
```

Let $T(n)$ be the running time of `sum_digits` where $n$ is the value of the input.

**(a)** How many digits does the number $n$ have, in terms of $n$? (Hint: think about logarithms.)

**(b)** Let $d$ be the number of digits in $n$. Write the running time equation $T(d)$ in terms of $d$, including the base case.

**(c)** Solve $T(d)$ by repeated substitution. What is the complexity class in terms of $d$? What is the complexity class in terms of $n$?

**Your Answer:**

**(a)** The number $n$ has $d = \lfloor \log_{10} n \rfloor + 1$ digits. Equivalently, $d = \Theta(\log n)$ (using any logarithm base, since they differ only by a constant factor).

**(b)** Each recursive call reduces the number of digits by 1 (since `n // 10` removes the last digit). The work done at each level is constant (one modulo operation, one addition). So:

$$T(1) = c \quad \text{(base case: single digit)}$$
$$T(d) = c + T(d - 1) \quad \text{for } d \ge 2$$

**(c)** Solving by repeated substitution:

$$T(d) = c + T(d-1)$$
$$= c + c + T(d-2) = 2c + T(d-2)$$
$$= 3c + T(d-3)$$
$$\vdots$$
$$= kc + T(d-k)$$

Setting $d - k = 1$ gives $k = d - 1$:
$$T(d) = (d-1)c + T(1) = (d-1)c + c = cd$$

So $T(d) = cd \in \Theta(d)$.

In terms of $n$: since $d = \Theta(\log n)$, the complexity is $\Theta(\log n)$.

## Question 2 (Non-coding)

Consider the following function:

```python
def mystery(L):

        return L
    mid = len(L) // 2
    left = mystery(L[:mid])
    right = mystery(L[mid:])
    return left + right
```

Recall that `L[:mid]` creates a new list of `mid` elements (costs `mid` steps), `L[mid:]` creates a new list of `n - mid` elements (costs `n - mid` steps), and `left + right` concatenates two lists (costs `len(left) + len(right)` steps).

**(a)** What does this function actually do? (Describe its output, not how it works.)

**(b)** Write the running time equation $T(n)$ for this function, where $n$ is `len(L)`. Account for the cost of slicing and concatenation.

**(c)** Solve the recurrence by repeated substitution. What is the complexity class?

**Your Answer:**

**(a)** The function returns a copy of the original list `L` (in the same order). It recursively splits the list in half and concatenates the halves back together, ultimately reconstructing the original list.

**(b)** Let $n$ = `len(L)`. The cost of `L[:mid]` is $\lfloor n/2 \rfloor$ steps, the cost of `L[mid:]` is $\lceil n/2 \rceil$ steps, and the cost of `left + right` is $n$ steps (since the two halves together have $n$ elements). The total non-recursive work is $n + n = 2n$ (slicing costs $n$ total, concatenation costs $n$). So:

$$T(1) = c$$
$$T(n) = 2n + 2T(n/2) \quad \text{for } n \ge 2$$

(We simplify by treating $\lfloor n/2 \rfloor \approx n/2$.)

**(c)** Solving by repeated substitution:

$$T(n) = 2n + 2T(n/2)$$
$$= 2n + 2[2(n/2) + 2T(n/4)] = 2n + 2n + 4T(n/4)$$
$$= 2n + 2n + 2n + 8T(n/8) \quad \text{(substituting again)}$$
$$\vdots$$
$$= 2nk + 2^k T(n/2^k)$$

Setting $n/2^k = 1$ gives $k = \log_2 n$:
$$T(n) = 2n \log_2 n + 2^{\log_2 n} \cdot c = 2n \log n + cn$$

Therefore $T(n) \in \Theta(n \log n)$.

## Question 4 (Non-coding)

Consider two functions that both compute the sum of all elements in a list:

**Version A (linear recursion):**
```python
def sum_A(L):
    if L == []:
        return 0
    return L[0] + sum_A(L[1:])
```

**Version B (divide and conquer):**
```python
def sum_B(L):
    if len(L) == 0:
        return 0
    if len(L) == 1:
        return L[0]
    mid = len(L) // 2
    return sum_B(L[:mid]) + sum_B(L[mid:])
```

Recall that `L[1:]` costs $n - 1$ steps and `L[:mid]` costs `mid` steps.

**(a)** Write the running time equation for `sum_A`.

**(b)** Write the running time equation for `sum_B`.

**(c)** Solve both recurrences. Which version is asymptotically faster?

**(d)** In merge sort, the recurrence $T(n) = n + 2T(n/2)$ gives us $\Theta(n \log n)$. Explain why `sum_B` has the same recurrence as merge sort but does **not** achieve $\Theta(n \log n)$, or explain why it does.

**Your Answer:**

**(a)** For `sum_A`: The slice `L[1:]` costs $n-1$ steps. The rest is constant work. So:

$$T(0) = c$$
$$T(n) = (n-1) + c + T(n-1) \quad \text{for } n \ge 1$$

Simplifying: $T(n) = n + T(n-1)$ (absorbing constants).

**(b)** For `sum_B`: The slice `L[:mid]` costs $n/2$ steps, `L[mid:]` costs $n/2$ steps. Total slicing cost is $n$. The addition is constant. So:

$$T(0) = c, \quad T(1) = c$$
$$T(n) = n + 2T(n/2) \quad \text{for } n \ge 2$$

**(c)** Solving `sum_A`:

$T(n) = n + T(n-1) = n + (n-1) + T(n-2) = \cdots = n + (n-1) + \cdots + 1 + T(0) = \frac{n(n+1)}{2} + c$

So `sum_A` is $\Theta(n^2)$.

Solving `sum_B`:

$T(n) = n + 2T(n/2)$. This is the same recurrence as merge sort. Solving:

$T(n) = n + 2[n/2 + 2T(n/4)] = 2n + 4T(n/4) = \cdots = kn + 2^k T(n/2^k)$

Setting $n/2^k = 1$: $T(n) = n\log n + cn \in \Theta(n \log n)$.

**`sum_B` ($\Theta(n \log n)$) is asymptotically faster than `sum_A` ($\Theta(n^2)$).**

**(d)** `sum_B` has the same recurrence form $T(n) = n + 2T(n/2)$ as merge sort, so it **does** achieve $\Theta(n \log n)$, not better. The $n$ term in the recurrence comes from the cost of slicing (creating new sublists), which is analogous to the merging step in merge sort. Even though `sum_B` doesn't do any "useful" combining work (it just adds two numbers), the cost of creating the sublists via slicing forces the same $\Theta(n \log n)$ running time. Both versions are slower than a simple loop ($\Theta(n)$) because of the overhead of list slicing.

## Question 5 (Coding)

Write a function `count_dc(L, x)` that counts how many times `x` appears in list `L` using a **divide-and-conquer** strategy.

**Strategy:** Split the list into two halves, recursively count occurrences in each half, and add the results.

Do **not** use loops. Use divide and conquer (split the list roughly in half), not linear recursion.

Example:
```python
count_dc([3, 1, 4, 1, 5, 1, 3], 1)   # returns 3
count_dc([2, 2, 2, 2], 2)              # returns 4
count_dc([1, 2, 3], 5)                 # returns 0
count_dc([], 1)                        # returns 0
```

````python
def count_dc(L, x):
    if len(L) == 0:
        return 0
    if len(L) == 1:
        if L[0] == x:
            return 1
        else:
            return 0
    mid = len(L) // 2
    return count_dc(L[:mid], x) + count_dc(L[mid:], x)

# Test your function
print(count_dc([3, 1, 4, 1, 5, 1, 3], 1))   # Expected: 3
print(count_dc([2, 2, 2, 2], 2))              # Expected: 4
print(count_dc([1, 2, 3], 5))                 # Expected: 0
print(count_dc([], 1))                        # Expected: 0
print(count_dc([7], 7))                       # Expected: 1
````

In Assignment 2, you wrote a **linear recursive** function `is_sorted(L)`. Now, write a **divide-and-conquer** version.

Write a function `is_sorted_dc(L)` that returns `True` if the list is in non-decreasing order.

**Strategy:** Split the list into two halves. The list is sorted if and only if:
1. The left half is sorted.
2. The right half is sorted.
3. The last element of the left half is $\le$ the first element of the right half.

Do **not** use loops.

Example:
```python
is_sorted_dc([1, 3, 5, 7, 9])     # returns True
is_sorted_dc([1, 3, 2, 7, 9])     # returns False
is_sorted_dc([5])                  # returns True
is_sorted_dc([])                   # returns True
is_sorted_dc([1, 1, 2, 3, 3])     # returns True
```

````python
def is_sorted_dc(L):
    if len(L) <= 1:
        return True
    mid = len(L) // 2
    if not is_sorted_dc(L[:mid]):
        return False
    if not is_sorted_dc(L[mid:]):
        return False
    # Check that last element of left half <= first element of right half
    if L[mid - 1] > L[mid]:
        return False
    return True

# Test your function
print(is_sorted_dc([1, 3, 5, 7, 9]))     # Expected: True
print(is_sorted_dc([1, 3, 2, 7, 9]))     # Expected: False
print(is_sorted_dc([5]))                  # Expected: True
print(is_sorted_dc([]))                   # Expected: True
print(is_sorted_dc([1, 1, 2, 3, 3]))     # Expected: True
print(is_sorted_dc([5, 4, 3, 2, 1]))     # Expected: False
print(is_sorted_dc([1, 2, 3, 1]))        # Expected: False
````

## Question 7 (Coding + Analysis)

Write a function `min_max(L)` that returns a list `[minimum, maximum]` of a non-empty list `L` using divide and conquer.

**Strategy:** Split the list in half. Recursively find the min and max of each half. Then combine: the overall min is the smaller of the two mins, and the overall max is the larger of the two maxes.

Do **not** use built-in `min()` or `max()` functions. Do **not** use loops.

Example:
```python
min_max([3, 1, 4, 1, 5, 9, 2, 6])  # returns [1, 9]
min_max([42])                        # returns [42, 42]
min_max([7, 3])                      # returns [3, 7]
```

````python
def min_max(L):
    if len(L) == 1:
        return [L[0], L[0]]
    if len(L) == 2:
        if L[0] < L[1]:
            return [L[0], L[1]]
        else:
            return [L[1], L[0]]
    mid = len(L) // 2
    left_mm = min_max(L[:mid])
    right_mm = min_max(L[mid:])
    # Combine: pick smaller min and larger max
    if left_mm[0] < right_mm[0]:
        overall_min = left_mm[0]
    else:
        overall_min = right_mm[0]
    if left_mm[1] > right_mm[1]:
        overall_max = left_mm[1]
    else:
        overall_max = right_mm[1]
    return [overall_min, overall_max]

# Test your function
print(min_max([3, 1, 4, 1, 5, 9, 2, 6]))  # Expected: [1, 9]
print(min_max([42]))                        # Expected: [42, 42]
print(min_max([7, 3]))                      # Expected: [3, 7]
print(min_max([5, 5, 5]))                   # Expected: [5, 5]
print(min_max([10, 1, 10, 1]))              # Expected: [1, 10]
````

**Analysis:** Write the running time equation for your `min_max` function (ignore the cost of slicing). What is the complexity class? Is this faster, slower, or the same as finding the min and max separately with two passes through the list?

**Your Answer:**

Ignoring the cost of slicing, the non-recursive work at each level is constant (two comparisons to combine the results). So:

$$T(1) = c, \quad T(2) = c$$
$$T(n) = c + 2T(n/2)$$

Solving: $T(n) = c + 2[c + 2T(n/4)] = 3c + 4T(n/4) = \cdots = c(2^k - 1) + 2^k T(n/2^k)$

Setting $n/2^k = 1$ ($k = \log_2 n$): $T(n) = c(n - 1) + cn = 2cn - c \in \Theta(n)$.

The complexity is $\Theta(n)$, which is the **same** as finding the min and max separately with two passes ($\Theta(n) + \Theta(n) = \Theta(n)$). However, the divide-and-conquer approach uses approximately $3n/2$ comparisons (compared to $2(n-1)$ comparisons for two separate passes), so it uses fewer comparisons in practice.

## Question 8 (Coding)

In class, we wrote functions to count nodes and count nodes of a given color in a binary tree. Now, write a function `all_values(tree)` that returns a **list** of all node colors in the tree.

Use the `BTree` class provided below (same as in class).

**Strategy:** If the tree is `None`, return an empty list. Otherwise, combine the current node's color with the values collected from the left and right subtrees.

Example (output will vary depending on the random tree):
```python
# If the tree looks like:
#   R
#     B
#       R
# Then all_values returns ['Red', 'Blue', 'Red'] (or similar order)
```

````python
import random

class BTree:
    def __init__(self, m=None):
        if m is None:
            m = random.randint(2, 7)
        self.left = None
        self.right = None
        if random.randint(0, 1) == 0:
            self.color = 'Blue'
        else:
            self.color = 'Red'
        if random.randint(0, m) > 0:
            self.left = BTree(m-1)
        if random.randint(0, m) > 0:
            self.right = BTree(m-1)

    def print_tree(self, indents=0):
        print('\t'*indents + self.color[0])
        if self.left is not None:
            self.left.print_tree(indents+1)
        if self.right is not None:
            self.right.print_tree(indents+1)
````

````python
def all_values(tree):
    if tree is None:
        return []
    return [tree.color] + all_values(tree.left) + all_values(tree.right)

# Test your function
random.seed(42)  # for reproducibility
t = BTree(3)
t.print_tree()
print("All values:", all_values(t))
print("Count check:", len(all_values(t)), "values collected")
````

**Explain your approach:** Describe your recursive strategy in English. Focus on *what* the function does more than *how*. What is the base case? How does each recursive call make progress toward the base case?

**Your Answer:**

**Base case:** If the tree is `None` (empty), return an empty list `[]`. There are no nodes, so there are no values to collect.

**Recursive case:** If the tree is not `None`, the current node has a color. We collect:
1. The current node's color as a one-element list `[tree.color]`.
2. All values from the left subtree (by recursively calling `all_values(tree.left)`).
3. All values from the right subtree (by recursively calling `all_values(tree.right)`).

We concatenate these three lists and return the result.

Each recursive call operates on a subtree, which has strictly fewer nodes than the current tree. Eventually, every path reaches a `None` child (the base case). This guarantees that every node in the tree is visited exactly once, and its color is included in the output list.

## Question 9 (Coding + Analysis)

In class, we implemented quick sort using the first element as the pivot. The `rearrange` function partitions the remaining elements into two lists: elements less than the pivot, and elements greater than or equal to the pivot.

Now, write a variation called `quick_sort_3way(L)` that partitions the list into **three** groups around the pivot (the first element):
- `less`: elements strictly less than the pivot
- `equal`: elements equal to the pivot (including the pivot itself)
- `greater`: elements strictly greater than the pivot

Then recursively sort only `less` and `greater` (no need to sort `equal`).

**Hint:** You may write a helper function `rearrange_3way(L)` that returns three lists. You may use loops inside the helper.

Example:
```python
quick_sort_3way([3, 1, 4, 1, 5, 9, 2, 6, 5, 3, 5])  # returns [1, 1, 2, 3, 3, 4, 5, 5, 5, 6, 9]
quick_sort_3way([5, 5, 5])                              # returns [5, 5, 5]
quick_sort_3way([])                                     # returns []
```

````python
def rearrange_3way(L):
    pivot = L[0]
    less = []
    equal = []
    greater = []
    for item in L:
        if item < pivot:
            less.append(item)
        elif item == pivot:
            equal.append(item)
        else:
            greater.append(item)
    return less, equal, greater

def quick_sort_3way(L):
    if len(L) <= 1:
        return L
    less, equal, greater = rearrange_3way(L)
    return quick_sort_3way(less) + equal + quick_sort_3way(greater)

# Test your function
print(quick_sort_3way([3, 1, 4, 1, 5, 9, 2, 6, 5, 3, 5]))  # Expected: [1, 1, 2, 3, 3, 4, 5, 5, 5, 6, 9]
print(quick_sort_3way([5, 5, 5]))                              # Expected: [5, 5, 5]
print(quick_sort_3way([]))                                     # Expected: []
print(quick_sort_3way([1]))                                    # Expected: [1]
print(quick_sort_3way([4, 2, 2, 8, 3, 3, 1]))                # Expected: [1, 2, 2, 3, 3, 4, 8]
````

**Analysis:** Consider the case where all elements in the list are the same (e.g., `[5, 5, 5, 5, 5]`).

**(a)** What happens when the `quick_sort` from class (with the 2-way `rearrange`) is called on this input? How many recursive calls are made?

**(b)** What happens when your `quick_sort_3way` is called on this input? How many recursive calls are made?

**(c)** Explain why the 3-way partition is an improvement for inputs with many duplicate values.

**Your Answer:**

**(a)** With the 2-way `rearrange` from class (using the first element as pivot), when all elements are the same (e.g., `[5, 5, 5, 5, 5]`), the pivot is 5. All remaining elements are $\ge$ the pivot, so they all go into the "greater than or equal" group. The "less than" group is empty. This means one recursive call gets an empty list and the other gets $n-1$ elements. This leads to $n - 1$ recursive calls (each removing only one element), giving $\Theta(n^2)$ behavior --- the worst case for quick sort.

**(b)** With `quick_sort_3way`, when all elements are the same, the `rearrange_3way` function puts all $n$ elements into the `equal` group. Both `less` and `greater` are empty. So both recursive calls are on empty lists and return immediately. Only **1** call to `rearrange_3way` is made (plus 2 trivial recursive calls on empty lists). The total work is $\Theta(n)$ for this case.

**(c)** The 3-way partition is an improvement for inputs with many duplicates because all copies of the pivot value are identified in a single pass and placed into the `equal` group. These elements never appear in any subsequent recursive call. In contrast, the 2-way partition only removes one copy of the pivot per recursive call, so $k$ copies of the same value require $k$ levels of recursion. With many duplicates, the 3-way version eliminates large chunks of the input at once, dramatically reducing the number of recursive calls.

## Question 10 (Non-coding)

Consider the `count_nodes` function from class:

```python
def count_nodes(t):
    if t is None:
        return 0
    return 1 + count_nodes(t.left) + count_nodes(t.right)
```

Let $T(n)$ be the running time of `count_nodes` where $n$ is the total number of nodes in the tree.

**(a)** In the worst case, how are the $n - 1$ remaining nodes distributed between `t.left` and `t.right`? In the best case (for analysis), how are they distributed?

**(b)** For a **balanced** binary tree (where the left and right subtrees each have roughly $n/2$ nodes), write the running time equation.

**(c)** Solve the recurrence from part (b) by repeated substitution and state the complexity class.

**(d)** Does the shape of the tree (balanced vs. unbalanced) affect the overall complexity of `count_nodes`? Explain why or why not. (Hint: think about how many times each node is visited, regardless of tree shape.)

**Your Answer:**

**(a)** In the **worst case**, all $n - 1$ remaining nodes are in one subtree and the other subtree is empty. This happens in a completely skewed tree (like a linked list). For example, `t.left` has $n - 1$ nodes and `t.right` has 0 nodes.

In the **best case** (for analysis), the tree is balanced: the remaining $n - 1$ nodes are split roughly evenly, with $\lfloor (n-1)/2 \rfloor$ nodes on one side and $\lceil (n-1)/2 \rceil$ on the other (approximately $n/2$ each).

**(b)** For a balanced binary tree:

$$T(0) = c \quad \text{(base case: } t \text{ is None)}$$
$$T(n) = c + T(n/2) + T(n/2) = c + 2T(n/2)$$

**(c)** Solving by repeated substitution:

$$T(n) = c + 2T(n/2)$$
$$= c + 2[c + 2T(n/4)] = 3c + 4T(n/4)$$
$$= 3c + 4[c + 2T(n/8)] = 7c + 8T(n/8)$$

General pattern after $k$ substitutions:
$$T(n) = (2^k - 1)c + 2^k T(n/2^k)$$

Setting $n/2^k = 0$ (i.e., $2^k = n$, so $k = \log_2 n$):
$$T(n) = (n - 1)c + n \cdot c = (2n - 1)c \in \Theta(n)$$

Complexity class: $\Theta(n)$.

**(d)** No, the shape of the tree does **not** affect the overall complexity. Regardless of whether the tree is balanced or skewed, `count_nodes` visits **every node exactly once** and does $\Theta(1)$ work at each node (one addition, two recursive calls). It also visits every `None` child exactly once (there are $n + 1$ `None` pointers in a binary tree with $n$ nodes). The total work is therefore $\Theta(n)$ for *any* tree shape.

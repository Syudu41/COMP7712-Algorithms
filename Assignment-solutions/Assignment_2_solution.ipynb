## Question 1 (Non-coding)

Consider the following recursive function:

```python
def mystery(base, exp):
    if exp == 0:
        return 1
    return base * mystery(base, exp - 1)
```

**(a)** Explain **what** this function does (not how it works, but what it accomplishes).

**(b)** Explain **how** this function accomplishes its task (describe the recursive strategy).

**(c)** Trace the sequence of calls and return values for `mystery(2, 4)`. Show each call and what it returns.

**Your Answer:**

(a) What: `mystery` computes `base` raised to the power of `exp` (i.e., base^exp).

(b) How: It uses the fact that base^exp = base * base^(exp-1). It reduces the exponent by 1 in each recursive call. When the exponent reaches 0, it returns 1 (since anything raised to the power 0 is 1).

(c) Trace:
```
mystery(2, 4)
  = 2 * mystery(2, 3)
  = 2 * (2 * mystery(2, 2))
  = 2 * (2 * (2 * mystery(2, 1)))
  = 2 * (2 * (2 * (2 * mystery(2, 0))))
  = 2 * (2 * (2 * (2 * 1)))
  = 2 * (2 * (2 * 2))
  = 2 * (2 * 4)
  = 2 * 8
  = 16
```

## Question 2 (Coding)

Write a recursive function `sum_list(L)` that computes the sum of all numbers in a list.

**Strategy:** The sum of a list is the first element plus the sum of the remaining elements.

Example:
```python
sum_list([1, 2, 3, 4, 5])  # returns 15
sum_list([10])              # returns 10
sum_list([])                # returns 0
```

````python
def sum_list(L):
    if L == []:
        return 0
    return L[0] + sum_list(L[1:])

# Test your function
print(sum_list([1, 2, 3, 4, 5]))  # Expected: 15
print(sum_list([10]))              # Expected: 10
print(sum_list([]))                # Expected: 0
print(sum_list([-1, 1, -2, 2]))   # Expected: 0
````

## Question 3 (Coding)

Write a recursive function `remove_all(L, x)` that returns a new list with all occurrences of `x` removed.

**Strategy:** Look at the first element. If it equals `x`, skip it and **remove x from the remaining elements**. Otherwise, keep it and **remove x from the remaining elements**.

Example:
```python
remove_all([1, 2, 3, 2, 4, 2], 2)  # returns [1, 3, 4]
remove_all([5, 5, 5], 5)            # returns []
remove_all([1, 2, 3], 4)            # returns [1, 2, 3]
```

````python
def remove_all(L, x):
    if L == []:
        return []
    if L[0] == x:
        return remove_all(L[1:], x)
    else:
        return [L[0]] + remove_all(L[1:], x)

# Test your function
print(remove_all([1, 2, 3, 2, 4, 2], 2))  # Expected: [1, 3, 4]
print(remove_all([5, 5, 5], 5))            # Expected: []
print(remove_all([1, 2, 3], 4))            # Expected: [1, 2, 3]
print(remove_all([], 1))                   # Expected: []
````

**Function 1:**
```python
def f(n):
    if n <= 1:
        return n
    return f(n - 1) + f(n - 2)
```

**Function 2:**
```python
def g(L):
    if L == []:
        return True
    if L[0] < 0:
        return False
    return g(L[1:])
```

**Your Answer:**

**Function 1:** (`f` computes the nth Fibonacci number)

(a) Base case(s): When `n <= 1`, it returns `n` directly (so `f(0) = 0` and `f(1) = 1`).

(b) Recursive case(s): When `n > 1`, it returns `f(n-1) + f(n-2)` — the sum of two smaller Fibonacci numbers.

(c) How the problem size is reduced: Each recursive call decreases `n` by 1 or by 2.

(d) Why it terminates: Since `n` decreases by at least 1 in every recursive call and the base case catches all values of `n <= 1`, the function will always eventually reach the base case (assuming `n` is a non-negative integer).

**Function 2:** (`g` checks whether all numbers in a list are non-negative)

(a) Base case(s): There are two base cases: (1) when `L` is empty, it returns `True`; (2) when `L[0] < 0`, it returns `False`.

(b) Recursive case(s): When the list is not empty and the first element is non-negative, it recurses on `L[1:]`.

(c) How the problem size is reduced: `L[1:]` has one fewer element than `L`, so the list gets shorter by 1 each time.

(d) Why it terminates: The list shrinks by one element in each recursive call. Eventually, either a negative number is found (returning `False`) or the list becomes empty (returning `True`).

## Question 5 (Coding)

Write a recursive function `is_palindrome(L)` that returns `True` if a list reads the same forwards and backwards, and `False` otherwise.

**Strategy:** Compare the first and last elements. If they are equal, check whether the middle portion (everything except the first and last) is also a palindrome.

Example:
```python
is_palindrome(list("racecar"))  # returns True
is_palindrome([1, 2, 3, 2, 1]) # returns True
is_palindrome([1, 2, 3])       # returns False
is_palindrome([7])             # returns True
is_palindrome([])              # returns True
```

````python
def is_palindrome(L):
    if len(L) <= 1:
        return True
    if L[0] != L[-1]:
        return False
    return is_palindrome(L[1:-1])

# Test your function
print(is_palindrome(list("racecar")))  # Expected: True
print(is_palindrome([1, 2, 3, 2, 1])) # Expected: True
print(is_palindrome([1, 2, 3]))       # Expected: False
print(is_palindrome([7]))             # Expected: True
print(is_palindrome([]))              # Expected: True
````

## Question 6 (Coding)

Rewrite the `select` function **recursively** (without using a for loop or while loop).

The function should behave exactly like the iterative version from class:

```python
def select(L, condition):
    output = []
    for item in L:
        if condition(item):
            output.append(item)
    return output
```

**Hint:** Look at the first element. If it satisfies the condition, include it in the result along with the recursive selection of the rest. Otherwise, just return the recursive selection of the rest.

Example:
```python
recursive_select([1,2,3,4,5,6], lambda x: x % 2 == 0)  # returns [2, 4, 6]
recursive_select(["hi", "hello", "hey", "world"], lambda w: len(w) > 3)  # returns ['hello', 'world']
```

````python
def recursive_select(L, condition):
    if L == []:
        return []
    if condition(L[0]):
        # take first ele together with remaining elements that satisfy the condition
        return [L[0]] + recursive_select(L[1:], condition)
    else:
        return recursive_select(L[1:], condition)

# Test your function
print(recursive_select([1,2,3,4,5,6], lambda x: x % 2 == 0))  # Expected: [2, 4, 6]
print(recursive_select(["hi", "hello", "hey", "world"], lambda w: len(w) > 3))  # Expected: ['hello', 'world']
print(recursive_select([1, -2, 3, -4], lambda x: x > 0))  # Expected: [1, 3]
print(recursive_select([], lambda x: True))                # Expected: []
````

## Question 8 (Coding)

Write a recursive function `is_sorted(L)` that returns `True` if a list of numbers is in non-decreasing order, and `False` otherwise.

**Strategy:** Compare the first two elements. If the first is greater than the second, the list is not sorted. Otherwise, check if the rest of the list (starting from the second element) is sorted.

Example:
```python
is_sorted([1, 3, 5, 7])     # returns True
is_sorted([1, 5, 3, 7])     # returns False
is_sorted([2])              # returns True
is_sorted([])               # returns True
is_sorted([1, 1, 2, 3, 3]) # returns True
```

````python
def is_sorted(L):
    if len(L) <= 1:
        return True
    if L[0] > L[1]:
        return False
    return is_sorted(L[1:])

# Test your function
print(is_sorted([1, 3, 5, 7]))     # Expected: True
print(is_sorted([1, 5, 3, 7]))     # Expected: False
print(is_sorted([2]))              # Expected: True
print(is_sorted([]))               # Expected: True
print(is_sorted([1, 1, 2, 3, 3])) # Expected: True
print(is_sorted([5, 4, 3, 2, 1])) # Expected: False
````

## Question 9 (Coding)

Given the following student data, write a recursive function `find_top_scorer(students)` that returns the **name** of the student with the highest score.

Each student is represented as a list: `[name, score]`

**Do not use** the built-in `max()` function or any loops.

**Hint:** Adapt the recursive `find_max` strategy from class, but work with `[name, score]` pairs instead of plain numbers.

Example:
```python
students = [
    ["Alice", 92],
    ["Bob", 78],
    ["Charlie", 95],
    ["Diana", 88]
]
find_top_scorer(students)  # returns "Charlie"
```

````python

def find_best(students):
    """Returns the [name, score] pair with the highest score."""
    if len(students) == 1:
        return students[0]
    first = students[0]
    best_of_rest = find_best(students[1:])   # use the same procedure to find the best student in the remaining list
    if first[1] >= best_of_rest[1]:
        return first
    else:
        return best_of_rest

def find_top_scorer(students):
    return find_best(students)[0]

# Test your function
students1 = [
    ["Alice", 92],
    ["Bob", 78],
    ["Charlie", 95],
    ["Diana", 88]
]
print(find_top_scorer(students1))  # Expected: Charlie

students2 = [
    ["Eve", 100],
    ["Frank", 85]
]
print(find_top_scorer(students2))  # Expected: Eve

students3 = [["Grace", 77]]
print(find_top_scorer(students3))  # Expected: Grace
````

## Question 10 (Coding + Explanation)

Write a recursive function `flatten(L)` that takes a nested list and returns a flat list containing all the values.

A nested list is a list whose elements can be either individual values or other lists (which may themselves contain lists).

**Hint:** For each element, check if it is a list (using `isinstance(element, list)`). If it is, flatten it recursively. If it is not, it is a single value.

Example:
```python
flatten([1, [2, 3], [4, [5, 6]], 7])  # returns [1, 2, 3, 4, 5, 6, 7]
flatten([[1, 2], [3, 4]])              # returns [1, 2, 3, 4]
flatten([1, 2, 3])                     # returns [1, 2, 3]
flatten([])                            # returns []
```

````python
def flatten(L):
    if L == []:
        return []
    first = L[0]
    rest = flatten(L[1:])
    if isinstance(first, list):
        return flatten(first) + rest
    else:
        return [first] + rest

# Test your function
print(flatten([1, [2, 3], [4, [5, 6]], 7]))  # Expected: [1, 2, 3, 4, 5, 6, 7]
print(flatten([[1, 2], [3, 4]]))              # Expected: [1, 2, 3, 4]
print(flatten([1, 2, 3]))                     # Expected: [1, 2, 3]
print(flatten([]))                            # Expected: []
print(flatten([[[1]], [2, [3, [4]]]]))        # Expected: [1, 2, 3, 4]
````

**Explain your approach:** Describe your recursive strategy in English. What is the base case? How does each recursive call make progress toward the base case?


**Your Answer:**

The base case is an empty list, which returns an empty list (there is nothing to flatten).

For the recursive case, I look at the first element of the list. If it is a plain value (not a list), I put it in a one-element list and concatenate it with the flattened rest of the list. If the first element is itself a list, I flatten it recursively and concatenate the result with the flattened rest.

Each recursive call makes progress because `L[1:]` is one element shorter than `L`, so the outer list shrinks toward the empty base case. When the first element is a nested list, `flatten(first)` also makes progress because the nested list is a smaller structure than the original input.

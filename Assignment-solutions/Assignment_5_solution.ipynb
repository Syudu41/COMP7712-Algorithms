## Question 1 (Non-coding)

Use the Master Theorem to determine the complexity class of each recurrence. For each one, show:
- The values of $a$, $b$, and $d$
- The value of $\log_b a$
- Which case of the Master Theorem applies

**(a)** $T(n) = n + 4\,T(n/2)$

**(b)** $T(n) = n^2 + 4\,T(n/2)$

**(c)** $T(n) = n^3 + 8\,T(n/2)$

**(d)** $T(n) = n + T(n/3)$

**(e)** $T(n) = n^2 + 9\,T(n/3)$

**(f)** $T(n) = 5 + T(n-1)$. Can the Master Theorem be applied here? Explain why or why not.

**Solution:**

For each, $T(n)=n^d + a\,T(n/b)$. Compare $\log_b a$ with $d$.

**(a)** $T(n)=n+4T(n/2)$: $a=4,\ b=2,\ d=1$. $\log_2 4 = 2 > 1$, so Case 1: $T(n)=\Theta(n^{\log_2 4})=\Theta(n^2)$.

**(b)** $T(n)=n^2+4T(n/2)$: $a=4,\ b=2,\ d=2$. $\log_2 4 = 2 = d$, Case 2: $T(n)=\Theta(n^2\log n)$.

**(c)** $T(n)=n^3+8T(n/2)$: $a=8,\ b=2,\ d=3$. $\log_2 8 = 3 = d$, Case 2: $T(n)=\Theta(n^3\log n)$.

**(d)** $T(n)=n+T(n/3)$: $a=1,\ b=3,\ d=1$. $\log_3 1 = 0 < 1$, Case 3: $T(n)=\Theta(n)$.

**(e)** $T(n)=n^2+9T(n/3)$: $a=9,\ b=3,\ d=2$. $\log_3 9 = 2 = d$, Case 2: $T(n)=\Theta(n^2\log n)$.

**(f)** $T(n)=5+T(n-1)$: The Master Theorem **cannot** be applied here. The Master Theorem requires the subproblem to be of size $n/b$ (a constant fraction of $n$), but here the subproblem is of size $n-1$. By repeated substitution, $T(n)=\Theta(n)$.

def all_distinct_B(L):

## Question 4 (Non-coding)

Consider these two functions that both compute the dot product of two lists of equal length $n$:

`dot_product([a1, a2, ..., an], [b1, b2, ..., bn])` $= a_1 \cdot b_1 + a_2 \cdot b_2 + \cdots + a_n \cdot b_n$

**Version A (linear recursion):**
```python
def dot_A(A, B):
    if len(A) == 0:
        return 0
    return A[0] * B[0] + dot_A(A[1:], B[1:])
```

**Version B (divide and conquer):**
```python
def dot_B(A, B):
    if len(A) == 0:
        return 0
    if len(A) == 1:
        return A[0] * B[0]
    mid = len(A) // 2
    return dot_B(A[:mid], B[:mid]) + dot_B(A[mid:], B[mid:])
```

Recall that `A[1:]` costs $n - 1$ steps, `A[:mid]` costs `mid` steps, etc.

**(a)** Write the running time equation for Version A. Solve it by repeated substitution. What is its complexity class?

**(b)** Write the running time equation for Version B. Note that there are 4 slicing operations (two for `A`, two for `B`). Identify $a$, $b$, $d$ and use the Master Theorem. What is its complexity class?

**(c)** Which version is asymptotically faster? Explain the result — does divide and conquer help here?

**Solution:**

**(a)** Version A: Each call does $O(1)$ arithmetic plus two slices `A[1:]` and `B[1:]`, each costing $n-1$. So:
$$T(n) = (n-1) + (n-1) + T(n-1) = 2(n-1) + T(n-1)$$
By repeated substitution: $T(n)=2(n-1)+2(n-2)+\cdots+2(1)+T(0)=2\cdot\frac{n(n-1)}{2}=\Theta(n^2)$.

**(b)** Version B: Two recursive calls of size $n/2$, plus 4 slices each costing $n/2$, so non-recursive cost is $4\cdot(n/2)=2n=\Theta(n)$:
$$T(n) = n + 2T(n/2)$$
$a=2,\ b=2,\ d=1$. $\log_2 2 = 1 = d$, Case 2: $T(n)=\Theta(n\log n)$.

**(c)** Version B is asymptotically faster ($\Theta(n\log n)$ vs $\Theta(n^2)$). Divide and conquer **does** help here — but only because of the expensive Python slicing. The "real" dot product cost is $\Theta(n)$; D&C beats the linear-recursion version because it splits the slicing cost across the recursion tree more evenly.

## Question 6 (Coding + Analysis)

Write a function `power(base, exp)` that computes `base` raised to the power `exp` (a non-negative integer) using **divide and conquer**.

**Strategy** (fast exponentiation):
- Base case: `base^0 = 1`
- If `exp` is even: `base^exp = (base^(exp/2))^2`. Compute `base^(exp/2)` recursively **once**, then square it.
- If `exp` is odd: `base^exp = base * base^(exp-1)`.

Do **not** use the built-in `**` operator or `pow()`. Do **not** use loops.

Example:
```python
power(2, 10)    # returns 1024
power(3, 0)     # returns 1
power(5, 3)     # returns 125
power(2, 20)    # returns 1048576
```

````python
def power(base, exp):
    if exp == 0:
        return 1
    if exp % 2 == 0:
        half = power(base, exp // 2)
        return half * half
    else:
        return base * power(base, exp - 1)

# Test your function
print(power(2, 10))    # Expected: 1024
print(power(3, 0))     # Expected: 1
print(power(5, 3))     # Expected: 125
print(power(2, 20))    # Expected: 1048576
print(power(7, 1))     # Expected: 7
print(power(3, 13))    # Expected: 1594323
````

**Analysis:**

**(a)** A naive recursive approach computes `base^exp = base * base^(exp-1)`, making one recursive call that reduces `exp` by 1 each time. Write the running time equation for the naive approach and state its complexity in terms of `exp`.

**(b)** In your divide-and-conquer version, what happens in the even case? How many recursive calls are made, and how does `exp` shrink? Write the running time equation for the even case.

**(c)** Use the Master Theorem on the even-case recurrence. What is the complexity of your D&C `power` function in terms of `exp`?

**(d)** How do the two approaches compare? If `exp = 1000`, approximately how many multiplications does each version perform?

**Solution:**

**(a)** Naive: $T(\text{exp})=1+T(\text{exp}-1)$. By repeated substitution, $T(\text{exp})=\Theta(\text{exp})$ — linear in `exp`.

**(b)** In the even case, we make **one** recursive call on `exp/2`, then do one multiplication (the squaring):
$$T(\text{exp}) = 1 + T(\text{exp}/2)$$

**(c)** $a=1,\ b=2,\ d=0$. $\log_2 1 = 0 = d$, Case 2: $T(\text{exp})=\Theta(\log \text{exp})$. (The odd case adds at most one extra step before becoming even, so it doesn't change the asymptotic class.)

**(d)** For `exp = 1000`: the naive version performs about 1000 multiplications; the D&C version performs about $\log_2 1000 \approx 10$ multiplications (up to roughly $2\log_2 1000 \approx 20$ counting odd-case steps). D&C is dramatically faster.

Consider two functions that both reverse a list.

**Version A (linear recursion):**
```python
def reverse_A(L):
    if L == []:
        return []

```

Recall that `L[1:]` costs $n - 1$ steps and `list1 + list2` costs `len(list1) + len(list2)` steps.

**(a)** (Non-coding) Write the running time equation for Version A. Solve it by repeated substitution and state the complexity class.

**(b)** Now write a **divide-and-conquer** version called `reverse_B(L)` that splits the list in half, recursively reverses each half, and combines the results. Think carefully about the order when combining.

Do **not** use loops. Do **not** use the built-in `reversed()` or slice reversal `[::-1]`.

Example:
```python
reverse_B([1, 2, 3, 4, 5])   # returns [5, 4, 3, 2, 1]
reverse_B([10, 20])           # returns [20, 10]
reverse_B([42])               # returns [42]
reverse_B([])                 # returns []
```

**Your Answer for (a):**

Version A: each call costs $n-1$ for the slice `L[1:]` plus $n$ for the concatenation `... + [L[0]]` (concatenating a list of length $n-1$ with one of length 1). So the non-recursive cost is $\Theta(n)$:
$$T(n) = n + T(n-1)$$
By repeated substitution: $T(n)=n+(n-1)+(n-2)+\cdots+1=\frac{n(n+1)}{2}=\Theta(n^2)$.

````python
def reverse_B(L):
    n = len(L)
    if n <= 1:
        return L[:]  # return a copy (or L itself)
    mid = n // 2
    left_rev = reverse_B(L[:mid])
    right_rev = reverse_B(L[mid:])
    # The reversed right half comes first, then the reversed left half
    return right_rev + left_rev

# Test your function
print(reverse_B([1, 2, 3, 4, 5]))   # Expected: [5, 4, 3, 2, 1]
print(reverse_B([10, 20]))           # Expected: [20, 10]
print(reverse_B([42]))               # Expected: [42]
print(reverse_B([]))                 # Expected: []
print(reverse_B([1, 2, 3, 4]))      # Expected: [4, 3, 2, 1]
````

**(c)** Write the running time equation for `reverse_B`. Account for the cost of slicing and concatenation. Identify $a$, $b$, $d$ and use the Master Theorem to find the complexity class.

**(d)** Which version is asymptotically faster? Does divide and conquer help for this problem?

**Your Answer:**

**(c)** Each call slices `L[:mid]` and `L[mid:]` (cost $n$ total) and concatenates two lists of length $n/2$ (cost $n$). So non-recursive work is $\Theta(n)$:
$$T(n) = n + 2T(n/2)$$
$a=2,\ b=2,\ d=1$. $\log_2 2 = 1 = d$, Case 2: $T(n)=\Theta(n\log n)$.

**(d)** Version B ($\Theta(n\log n)$) is asymptotically faster than Version A ($\Theta(n^2)$). Yes, divide and conquer helps — again, mostly because it amortizes the cost of Python's slicing/concatenation across the recursion tree.

## Question 8 (Coding + Analysis)

In Assignment 4, we worked with binary trees using the `BTree` class. Recall that each node has a `.color` (either `'Blue'` or `'Red'`), a `.left` subtree, and a `.right` subtree.

Write a function `tree_height(t)` that returns the **height** of a binary tree. The height is the length of the longest path from the root to a leaf, measured in edges (not nodes).

- An empty tree (`None`) has height $-1$.
- A single node (no children) has height $0$.
- A node with children has height $1 +$ the maximum of the heights of its subtrees.

Do **not** use loops.

Example:
```
# Tree:       R
#            / \
#           B   R
#          /
#         R
# Height = 2 (path: R -> B -> R)
```

````python
import random

class BTree:
    def __init__(self, m=None):
        if m is None:
            m = random.randint(2, 7)
        self.left = None
        self.right = None
        if random.randint(0, 1) == 0:
            self.color = 'Blue'
        else:
            self.color = 'Red'
        if random.randint(0, m) > 0:
            self.left = BTree(m-1)
        if random.randint(0, m) > 0:
            self.right = BTree(m-1)

    def print_tree(self, indents=0):
        print('\t'*indents + self.color[0])
        if self.left is not None:
            self.left.print_tree(indents+1)
        if self.right is not None:
            self.right.print_tree(indents+1)
````

````python
def tree_height(t):
    if t is None:
        return -1
    lh = tree_height(t.left)
    rh = tree_height(t.right)
    if lh > rh:
        return 1 + lh
    else:
        return 1 + rh

# Test your function
random.seed(42)
t = BTree(3)
t.print_tree()
print("Height:", tree_height(t))

# Additional tests
print("Height of None:", tree_height(None))  # Expected: -1

# Single node
single = BTree(0)
single.left = None
single.right = None
print("Height of single node:", tree_height(single))  # Expected: 0
````

**Analysis:**

**(a)** For a **balanced** binary tree with $n$ nodes (where the left and right subtrees each have roughly $n/2$ nodes), write the running time equation for `tree_height`.

**(b)** Use the Master Theorem to determine the complexity class.

**(c)** Does the shape of the tree (balanced vs. unbalanced) affect the overall complexity? Explain. (Hint: think about how many nodes are visited regardless of tree shape.)

**Solution:**

**(a)** For a balanced tree with $n$ nodes, each call does $O(1)$ work plus two recursive calls on subtrees of size $\approx n/2$:
$$T(n) = 1 + 2T(n/2)$$

**(b)** $a=2,\ b=2,\ d=0$. $\log_2 2 = 1 > 0$, Case 1: $T(n)=\Theta(n^{\log_2 2})=\Theta(n)$.

**(c)** No, the shape doesn't change the asymptotic complexity. Regardless of whether the tree is balanced or skewed, every node is visited exactly once and $O(1)$ work is done per node, so the total is always $\Theta(n)$.

## Question 9 (Coding + Analysis)

Write a function `count_color(t, color)` that counts the number of nodes in a binary tree with the given color. Then write a function `most_common_color(t)` that returns whichever color (`'Blue'` or `'Red'`) appears more often. If they are equal, return either one.

Use the `BTree` class from Question 8. Do **not** use loops.

Example:
```python
# If the tree has 5 Blue nodes and 3 Red nodes:
count_color(t, 'Blue')     # returns 5
count_color(t, 'Red')      # returns 3
most_common_color(t)       # returns 'Blue'
```

````python
def count_color(t, color):
    if t is None:
        return 0
    c = 1 if t.color == color else 0
    return c + count_color(t.left, color) + count_color(t.right, color)

def most_common_color(t):
    blue = count_color(t, 'Blue')
    red = count_color(t, 'Red')
    if blue >= red:
        return 'Blue'
    else:
        return 'Red'

# Test your function
random.seed(42)
t = BTree(4)
t.print_tree()
print("Blue count:", count_color(t, 'Blue'))
print("Red count:", count_color(t, 'Red'))
print("Most common:", most_common_color(t))
````

**Analysis:**

**(a)** Your `most_common_color` function calls `count_color` twice (once for Blue, once for Red). If `count_color` visits every node in the tree, what is the total running time of `most_common_color` in terms of $n$ (the number of nodes)?

**(b)** Now write an **improved** version called `most_common_color_v2(t)` that traverses the tree only **once** and returns the most common color. (Hint: write a helper that returns both counts simultaneously.)

````python
def _counts(t):
    if t is None:
        return [0, 0]
    lc = _counts(t.left)
    rc = _counts(t.right)
    blue = lc[0] + rc[0] + (1 if t.color == 'Blue' else 0)
    red  = lc[1] + rc[1] + (1 if t.color == 'Red'  else 0)
    return [blue, red]

def most_common_color_v2(t):
    blue, red = _counts(t)
    if blue >= red:
        return 'Blue'
    else:
        return 'Red'

# Test — should give same result as most_common_color
random.seed(42)
t = BTree(4)
print("Most common (v2):", most_common_color_v2(t))
````

**(c)** What is the running time of `most_common_color_v2`? Are the two versions in the same complexity class, or is one asymptotically faster?

**Solution:**

**(a)** `count_color` visits every node once, so it is $\Theta(n)$. Calling it twice gives $2\cdot\Theta(n)=\Theta(n)$ — still linear, but with a constant factor of 2.

**(b)** See `most_common_color_v2` above: a single helper traverses the tree once and returns both counts simultaneously.

**(c)** `most_common_color_v2` is also $\Theta(n)$, so the two versions are in the **same complexity class**. However, v2 is about twice as fast in practice because it traverses the tree only once instead of twice.

## Question 10 (Coding + Analysis)

Write a function `tree_mirror(t)` that returns a **new** tree that is the mirror image of `t`. In a mirrored tree, every node's left and right children are swapped (recursively).

Do **not** modify the original tree. Do **not** use loops.

Example:
```
# Original:          Mirror:
#      R                R
#     / \              / \
#    B   R            R   B
#   /     \          /     \
#  R       B        B       R
```

Hint: Create a new `BTree` node, set its color, and set its left and right children to the mirrored versions of the original's right and left children.

````python
def tree_mirror(t):
    if t is None:
        return None
    new = BTree.__new__(BTree)  # create without invoking random constructor
    new.color = t.color
    new.left  = tree_mirror(t.right)
    new.right = tree_mirror(t.left)
    return new

# Test your function
random.seed(42)
t = BTree(3)
print("Original:")
t.print_tree()

m = tree_mirror(t)
print("\nMirrored:")
m.print_tree()

print("\nOriginal unchanged:")
t.print_tree()
````

**Analysis:**

**(a)** For a balanced binary tree with $n$ nodes, write the running time equation for `tree_mirror`. (Hint: how much non-recursive work is done per call?)

**(b)** Use the Master Theorem to determine the complexity class.

**(c)** A student claims: "Mirroring a tree takes $O(n \log n)$ because it's like merge sort — we split in half and recombine." Is this correct? Explain why or why not.

**Solution:**

**(a)** Each call does $O(1)$ work (allocate a node, copy color, swap children) and makes two recursive calls on subtrees of size $\approx n/2$:
$$T(n) = 1 + 2T(n/2)$$

**(b)** $a=2,\ b=2,\ d=0$. $\log_2 2 = 1 > 0$, Case 1: $T(n)=\Theta(n)$.

**(c)** The student is **wrong**. Mirroring is not like merge sort because there is no expensive combine step — the "combine" is just attaching two pointers, which is $O(1)$, not $O(n)$. Merge sort is $\Theta(n\log n)$ because its merge step is $\Theta(n)$, giving $T(n)=n+2T(n/2)$. Here the per-node work is $O(1)$, so $T(n)=1+2T(n/2)=\Theta(n)$. Every node is visited exactly once.

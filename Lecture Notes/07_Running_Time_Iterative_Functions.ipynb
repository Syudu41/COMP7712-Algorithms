# Running Time of Iterative Programs

**Goal:** Determine the running time equation T(n) and its complexity.

- **T(n)** = number of steps the program takes as a function of input size n
- If a program sorts a list, n is the number of items in the list
- T(n) grows as n increases

### What is a Step?

A step is a basic unit of work (arithmetic, comparison, memory access) — an abstract measure independent of hardware. Counting steps lets us compare algorithms without worrying about implementation details.

```python
a = 2*b + 1   # 3 operations or 1 "step" — depends on abstraction level
```

````python
#
# Count the number of steps of this program 
#
def count_hello(words):          # T(n) = the number of steps count_hello takes
    count = 0                                  # constant
    for word in words:                         # n
        if word == 'hello':                    #    constant
            count += 1                         #    constant
    return count                               # constant
````

````python
count_hello(['hello', 'world', 'how', 'are', 'you'])
````

$n$ = number of items in `words`.

| Lines | Steps |
|-------|-------|
| 5 (init) | 1 |
| 6–8 (loop body) | $3 \times n$ |
| 9 (return) | 1 |

**Precise count:** $T(n) = 3n + 2$

### Constant Steps

A *constant step* is any operation taking fixed time regardless of $n$. Since exact operation counts are hardware-dependent, we group constants:

- Lines 5 + 9 = constant "a"
- Lines 7–8 = constant "b" per iteration

**Better way to describe the running time equation of this function:** $$T(n) = b \cdot n + a$$

### Nested Loops

````python
#
# Find T(n) assuming each operation takes 1 step. L has n items.
#
def foo(L):
    total = 0                          # 1
    for i,x in enumerate(L):           # n
        for j,y in enumerate(L):       #    n
            total += x*y               #       3
    return total                       # 1

# T(n) = 3n^2 + 2
# With constant steps: T(n) = a*n^2 + b
````

````python
foo([4,5,6])
````

Write the running time equation of this code:
```python
sum = 0                             # constant steps (a1)
for i in range(n):                  # n steps
    for j in range(i+1, n):         #    at most n steps
        sum += i*i                  #        constant steps (b)
print(sum)                          # constant steps (a2)
```

lines 7-8 take at most b*n steps.

lines 6-8 take at most b*n^2 steps.

$$T(n) \le a + b*n^2 $$


The inner loop runs from `i+1` to `n`, so we unroll:

| i | inner iterations |
|---|-----------------|
| 0 | $n-1$ |
| 1 | $n-2$ |
| ... | ... |
| $n-1$ | 0 |

Total = $(0 + 1 + \cdots + (n-1)) \cdot b = b \cdot \frac{n(n-1)}{2} = \frac{b \cdot n^2}{2} - \frac{b \cdot n}{2}$

$\frac{b \cdot n^2}{4} \le T(n) \le \frac{b \cdot n^2}{2}$

$$T(n) \le a + b*n*(n-i) $$

$a + b*n^2$ is an upper bound of $T(n)$.

$$c*n \le T(n) \le a+b*n^2$$

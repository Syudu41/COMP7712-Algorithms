# Running time of simple iterative programs

The goal of analyzing running time of a program is (1) to determine the running time equation, and (2) to determine the complexiy of the running time equation.

Describe the running time equation of a program.

Conventionally, the name of the equation is `T`. It has typically one parameter `n`, which is the size of the input.

`T(n)` is the running time equation of a program.

If a program sorts a list of items, what is `n`? `n` is the number of items in the list.


`T(n)` grows when `n` increases.

**What does `T(n)` measure? a number of steps the program take in relation to `n`.**


```python
a = 2*b + 1
```
How many "steps" are there in this statement? 3 steps or 1 step depending on what we want.


### What is a "step"?

A step is a basic unit of work performed by the algorithm. It is an abstract measure of computational effort, not tied to specific hardware or implementation details.

Examples of steps include:
+ Arithmetic operations (e.g., addition, multiplication).
+ Comparisons (e.g., checking if one number is greater than another).
+ Memory accesses (e.g., reading or writing a value).
+ The exact definition of a step depends on the algorithm and the level of abstraction being used.

Why Steps Matter:

+ Steps provide a hardware-independent way to analyze algorithms.
+ By counting steps, we can compare the efficiency of different algorithms without worrying about implementation details or machine-specific performance.


````python
#
# Count the number of steps of this program 
#
def count_hello(words):          # T(n) = the number of steps count_hello takes
    count = 0                                  # 1
    for word in words:                         # n
        if word == 'hello':                    # 1
            count += 1                         # 2
    return count                               # 1
````

````python
count_hello(['hello', 'world', 'how', 'are', 'you'])
````

### Constant Steps

A constant step refers to any operation or sequence of operations that takes a fixed amount of time, regardless of the input size. In other words, the time required to execute a constant step does not depend on `n`.

Why use constant steps:
+ Counting every single operation can be tedious and unnecessary
+ Hardware Independence


````python
#
# 
#
def count_hello(words):
    count = 0
    for word in words:
        if word == 'hello':
            count += 1
    return count
````

````python
#PID:7
#
# Find T(n) as precisely you can, assuming each operation takes 1 step.
#
# Suppose L has n items; n is the input size.
def foo(L):
    total = 0                          # 1
    for i,x in enumerate(L):           # n
        for j,y in enumerate(L):       #    n
            total += x*y               #       3
    return total                       # 1

# T(n) = 3n^2 + 2
# If different operations do not have the same number of instructions,
# then we say T(n) = a*n^2 + b
````

### Count steps involving loops

````python
n = 6
sum = 0
for i in range(n):        
    sum += i*i
print(sum)

sum = 0
for i in range(n):    
    for j in range(n):
        sum += i*i
print(sum)


sum = 0
for i in range(n):    
    for j in range(i+1, n):
        sum += i*i
print(sum)
````

lines 3-4: $n * a$ steps

lines 8-10: $n^2 * a$ steps

lines 15-17: at most $n^2 * a$ steps.  $T(n) \le n^2 * a$

We "unroll" the nested loop to get an exact count. This is a simple case.  Other complicated nested loops with conditions, etc., it's very hard to unroll the loop.

line 15 has n iterations. First, we unroll this outer loop.

+ i=0, lines 16-17, (n-1)*a
+ i=1, lines 16-17, (n-2)*a
+ i=2, lines 16-17, (n-3)*a
+ i=3, lines 16-17, (n-4)*a
...
+ i=n-1, lines 16-17, 0

Add all of these: $(0+1+2+ \cdots + (n-1))*a$

$T(n) = a * {(n-1)*n \over 2}$

$T(n) = a*n^2/2 - a*n/2$

$ a*n^2/4 \le T(n) \le a*n^2/2$

lower bound constant: $a/4$

upper bound constant: $a/2$

This shows that $T(n) \in \Theta(n^2)$

## The Master Theorem

Many divide-and-conquer algorithms follow this pattern:
- We solve a problem of size $n$ by breaking it into $a$ subproblems, each of size $n/b$.
- The non-recursive work (dividing and combining) takes $\Theta(n^d)$ time.

This theorem allows us to solve for T(n), where T(n) has this form:

$$T(n) = n^d + a \cdot T\left(\frac{n}{b}\right)$$

where:
- $a$ = number of recursive calls
- $b$ = factor by which input size shrinks.  n/b is the input size of each recursive call.
- $n^d$ = cost of non-recursive work

The running time has two components:
- iterative or non-recursive work: $n^d$
- recursive work: $a * T(n/b)$




### Understanding the parameters

**Example (merge sort)**:
- $a = 2$ (two recursive calls)
- $b = 2$ (each subproblem has size $n/2$)
- $d = 1$ (merging takes $\Theta(n)$ time, i.e. $n^1$)

So: $T(n) = n^1 + 2 \cdot T(n/2)$

Ultimate goal: given a problem, you can quickly design a correct strategy that is efficient.

- We spent a lot of time learning about how to design correct solutions: mathematical induction, mastering recursive designs, etc.
- Now, we will learn how to quickly figure out the running time complexity of these types of recursive desgins.

### Statement of the Master Theorem

Given $T(n) = n^d + a \cdot T\left(\frac{n}{b}\right)$, compare $d$ and $\log_b a$:

1. If $d \neq \log_b a$, then $T(n) \in \Theta\left(n^{\max(d,\; \log_b a)}\right)$

2. If $d = \log_b a$, then $T(n) \in \Theta(n^d \log n)$

$d$ tells us about the non-recursive calculations.

$\log_b a$ tells us about the recursive calculations.

If we use substitution, we will figure out the general formula to quick determine the complexity of $T(n)$.

---

**Example 1: $T(n) = n + 2*T({n \over 2})$**

d = 1

log_b (a) = log_2 (2) = 1

$$T(n) \in \Theta(n \log n)$$

**Example 2: $T(n) = n^2 + 2*T({n \over 2})$**  For example, we merge inefficiently.

d = 2

log_b(a) = log_2(2) = 1


$$T(n) \in \Theta(n^2)$$

**Example 3: $T(n) = n + 3*T({n \over 2})$**  For example, we recursve inefficiently.

d=1

log_b(a) = log_2(3) > 1

$$T(n) \in \Theta(n^{\log_2(3)})$$

### Useful facts about logarithms

- $\log_b a = d$ means $b^d = a$
- $\log_b b^x = x$
- $\log_b(a \cdot c) = \log_b a + \log_b c$

Quick estimate: $\log_2 32 = 5$ and $\log_2 64 = 6$, so $\log_2 50$ is between 5 and 6.

## Applying the Master Theorem: Examples

### Example 1: $T(n) = n + 2T(n/2)$ (merge sort)

- $a = 2$, $b = 2$, $d = 1$
- $\log_b a = \log_2 2 = 1$
- Since $d = \log_b a = 1$, we use case 2:

$$T(n) \in \Theta(n^1 \log n) = \Theta(n \log n)$$

### Example 2: $T(n) = 4 + T(n/4)$ (binary-search-like)

We can write this as $T(n) = n^0 + 1 \cdot T(n/4)$, since constant = $n^0$.

- $a = 1$, $b = 4$, $d = 0$
- $\log_b a = \log_4 1 = 0$
- Since $d = \log_b a = 0$, we use case 2:

$$T(n) \in \Theta(n^0 \log n) = \Theta(\log n)$$

### Example 3: $T(n) = n + 4T(n/2)$

- $a = 4$, $b = 2$, $d = 1$
- $\log_b a = \log_2 4 = 2$
- Since $d = 1 \neq 2 = \log_b a$, we use case 1:

$$T(n) \in \Theta\left(n^{\max(1, 2)}\right) = \Theta(n^2)$$

### Example 4: $T(n) = n^2 + 2T(n/2)$

- $a = 2$, $b = 2$, $d = 2$
- $\log_b a = \log_2 2 = 1$
- Since $d = 2 \neq 1 = \log_b a$, we use case 1:

$$T(n) \in \Theta\left(n^{\max(2, 1)}\right) = \Theta(n^2)$$

### Example 5: $T(n) = 5n^3 + 8T(n/2)$

- $a = 8$, $b = 2$, $d = 3$
- $\log_b a = \log_2 8 = 3$
- Since $d = \log_b a = 3$, we use case 2:

$$T(n) \in \Theta(n^3 \log n)$$

---

## Important: When the Master Theorem Does NOT Apply

The Master Theorem requires that each recursive call reduces the input by a **constant factor** ($n/b$).

It does **not** apply to recurrences like:
- $T(n) = c + T(n-1)$  (input shrinks by 1, not by a factor)
- $T(n) = n + T(n-1)$  (same issue)
- $T(n) = c + 2T(n-1)$ (same issue)

For these, we use repeated substitution (as we did in Cases 1, 2, and 4 above).

```python
def search(L, x, left, right):
    if left > right:
        return False
    mid = (left + right) // 2
    if x == L[mid]:
        return True
    elif x < L[mid]:
        return search(L, x, left, mid - 1)
    else:
        return search(L, x, mid + 1, right)
```

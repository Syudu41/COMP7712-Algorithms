# Program Correctness

## What Does It Mean for a Program to Be Correct?

A program is **correct** if it does what it is supposed to do for **all** valid inputs.

How can we convince ourselves that a program is correct?

- **Testing**: Run the program on a few inputs and check the output. But testing can only show the *presence* of bugs, not their *absence*. We cannot test all possible inputs.

- **Reasoning**: Argue logically that the program works for all inputs.

In this lecture, we will learn a reasoning technique for **recursive programs**.

---
## A Motivating Problem

Compute the sum of all items in a list, recursively.

````python
def sum_rec(L):
    if L == []:
        return 0
    first = L.pop(0)
    return first + sum_rec(L)

sum_rec([3, 1, 4, 2])
````

Let's trace what happens with `L = [3, 1, 4, 2]`:

```
sum_rec([3, 1, 4, 2])
  = 3 + sum_rec([1, 4, 2])        # relies on sum_rec([1, 4, 2]) being correct
  = 3 + (1 + sum_rec([4, 2]))     # relies on sum_rec([4, 2]) being correct
  = 3 + (1 + (4 + sum_rec([2])))  # relies on sum_rec([2]) being correct
  = 3 + (1 + (4 + (2 + sum_rec([]))))  # relies on sum_rec([]) being correct
  = 3 + (1 + (4 + (2 + 0)))
  = 10
```

**Observation**: Each call **relies on a correct answer for a smaller input**.

`sum_rec([3, 1, 4, 2])` does not solve the entire problem from scratch. It removes one element and **delegates the remaining work** to `sum_rec([1, 4, 2])`, trusting that call to return the correct answer.

---

## The Key Idea

Recursive programs achieve correctness by **building on correct solutions to smaller subproblems**.

If we can show two things:

1. The program is correct for the **smallest case** (base case).
2. **If** the program correctly handles a smaller subproblem, **then** it correctly handles the current problem.

Then the program is correct **for all inputs**.

But why do these two things guarantee correctness for *all* inputs? This is where **mathematical induction** comes in.

---

## Mathematical Induction

Mathematical induction is a proof technique that formalizes the reasoning we just used. Think of it like climbing a ladder:

- **Step 1** (Base case): You can stand on the first rung.
- **Step 2** (Inductive step): If you can stand on rung `k`, you can always step up to rung `k+1`.

If both are true, you can reach **any** rung.

Why? You can reach rung 1 (Step 1). Since you can reach rung 1, you can reach rung 2 (Step 2 with k=1). Since you can reach rung 2, you can reach rung 3 (Step 2 with k=2). And so on, forever.

### Formal Statement

Suppose we want to prove that a statement `P(n)` is true for all integers `n >= 0`. If we can prove:

1. **Base case**: `P(0)` is true.
2. **Inductive step**: For any `k >= 0`, if `P(k)` is true, then `P(k+1)` is also true.

Then `P(n)` is true for **all** `n >= 0`.

### A Quick Example (not about programs)

Prove that $1 + 2 + 3 + \cdots + n = \frac{n(n+1)}{2}$ for all $n \ge 1$.

We want to prove that $P(n)$, which is the statement "$1+2+\cdots+n = \frac{n(n+1)}{2}$", for all $n \ge 1$

Mathematical induction -- a proof technique -- has two steps:

1. Prove that $P(1)$ is true.  Is it true?   $P(1)$ is "$1 = \frac{1*2}{2}$".
   

Yes, P(1) is true.

Now, we need to prove the second step:  "If $P(k)$ is true, then $P(k+1)$ is also true for all $k>1$"

If $P(k)$ is true, then $1+2+\cdots+k = {k(k+1) \over 2}$.

Now, if this is true, then $1+2+\cdots+k+(k+1) = {k(k+1) \over 2} + (k+1) = {k(k+1) \over 2} + {2(k+1) \over 2} = {(k+1)(k+2) \over 2}$

So, $P(k+1)$ is true.

What is $P(k+1)$?  It is this statement: "$1 + 2 + \cdots + (k+1) = {(k+1)(k+2) \over 2}$".

We just showed that for $k>1$, if $P(k)$ is true, then $P(k+1)$ is true.  This is the second step of mathematical induction.

Therefore, $P(n)$ is true for all values of $n$.

### Inductive Reasoning vs. Mathematical Induction

**Inductive reasoning** (testing a few cases and concluding a general pattern) is **not** a proof.

- Testing P(1), P(2), P(3), P(10) and concluding "P(n) is true for all n" is inductive reasoning. It could be wrong.

**Mathematical induction** is a valid proof. It proves the statement for **all** values, not just the ones we tested.

## Applying Induction to Recursive Programs

In a recursive program, the function **calls itself with a smaller input**. Proving correctness follows the same inductive pattern:

1. **Base case**: The function is correct for the smallest input.
2. **Inductive step**: If the recursive call returns the correct answer for a smaller input, then the function returns the correct answer for the current input.

The recursive call **is** the assumption of the inductive step. When we write `return first + sum_rec(L)`, we are assuming that `sum_rec(L)` (with a smaller list) returns the correct answer, and building on it.

### Example: Proving `sum_rec` Correct

```python
def sum_rec(L):
    if L == []:
        return 0
    first = L.pop(0)
    return first + sum_rec(L)
```

We want to prove: "`sum_rec` correctly computes the sum of any input list."

Let `P(n)` be: "`sum_rec` is correct when the input list has `n` items."

**Base case** `P(0)`: When `L` is empty, `sum_rec` returns 0. The sum of an empty list is 0. Correct.

**Inductive step**: Assume `P(k)` is true for all `k < n`, i.e., `sum_rec` works correctly for any list with fewer than `n` items.

When the input has `n` items:
- `first` gets `L[0]`, and `L` now has `n-1` items.
- The recursive call `sum_rec(L)` receives a list of `n-1` items. By our assumption, it correctly returns the sum of those `n-1` items.
- The function returns `first + (sum of remaining items)` = sum of all `n` items. Correct.

By mathematical induction, `sum_rec` is correct for all input sizes.

```
reverse([1, 2, 3, 4])
  first = 1, then call reverse([2, 3, 4])
    first = 2, then call reverse([3, 4])
      first = 3, then call reverse([4])
        first = 4, then call reverse([])
          returns []
        returns [] + [4] = [4]
      returns [4] + [3] = [4, 3]
    returns [4, 3] + [2] = [4, 3, 2]
  returns [4, 3, 2] + [1] = [4, 3, 2, 1]
```

**Correctness argument**:

We want to prove: "`reverse` correctly reverses any input list."

**Inductive step**: Assume `reverse` works correctly for any list with fewer than `n` items.

When the input list is `[a, b, c, ..., z]` (n items):
- `first = a`, and `L` becomes `[b, c, ..., z]` (n-1 items).
- By our assumption, `reverse([b, c, ..., z])` correctly returns `[z, ..., c, b]`.
- The function returns `[z, ..., c, b] + [a]` = `[z, ..., c, b, a]`.
- This is the reverse of `[a, b, c, ..., z]`. Correct.

## Basic Induction vs. Strong Induction

There is one structural difference worth noting.

**Basic induction**: Assume `P(k)` is true. Prove `P(k+1)`.

This is natural for recursive programs that reduce the input by one (e.g., processing one element at a time).

**Strong induction**: Assume `P(0), P(1), ..., P(k)` are **all** true. Prove `P(k+1)`.

This is natural for recursive programs where the recursive call might reduce the input by more than one (e.g., splitting a list in half).

Strong induction is at least as powerful as basic induction. Anything you can prove with basic induction, you can also prove with strong induction.

In terms of program correctness, the program is correct is we can prove 2 things:

1. It is correct for the smallest inputs.

2. If the recursive call(s) have smaller inputs and all return correct answers, then based on this assumption, the program is correct.

If we can prove these 2 steps, then the program is correct for all inputs.

### Example: Finding Min and Max (Divide and Conquer)

This recursive function splits the list in half. We need strong induction because the recursive calls are on inputs of size roughly `n/2`, not `n-1`.

````python
def find_min_max(L):
    if len(L) == 1:
        return L[0], L[0]
    if len(L) == 2:
        if L[0] < L[1]:
            return L[0], L[1]
        else:
            return L[1], L[0]
    mid = len(L) // 2
    left_min, left_max = find_min_max(L[:mid])
    right_min, right_max = find_min_max(L[mid:])
    return min(left_min, right_min), max(left_max, right_max)

find_min_max([10, 9, 26, 7, 5, 20, 15, 2, 8])
````

**Correctness argument**:

**Base cases**: A list of 1 element: the element is both the min and max. A list of 2 elements: compare them. Both are correct.

**Inductive step**: Assume `find_min_max` is correct for **all** lists with fewer than `n` items (strong induction).

When the input has `n` items:
- The list is split into two halves, each with fewer than `n` items.
- By our assumption, `find_min_max` correctly returns the min and max of each half.
- The overall min is the smaller of the two half-mins; the overall max is the larger of the two half-maxes.
- This gives the correct min and max of the entire list.

Note: We need **strong** induction here because the recursive calls are on inputs of size `n/2`, not `n-1`. We need to assume correctness for all smaller sizes, not just size `n-1`.

---

## Summary

Mathematical induction formalizes the reasoning behind recursive correctness:

| Induction | Recursive Program |
|:---|:---|
| Base case: P(0) is true | Function is correct for the smallest input (e.g., empty list) |
| Inductive step: P(k) implies P(k+1) | If the recursive call is correct, the function is correct |
| Conclusion: P(n) for all n | The function is correct for all input sizes |

The recursive call **is** the inductive hypothesis made concrete. When we trust `sum_rec(L)` to return the correct answer for a smaller list, we are applying the inductive assumption.

- **Basic induction** suffices when recursion reduces input by one (linear recursion).
- **Strong induction** is needed when recursion reduces input by more than one (e.g., divide and conquer).

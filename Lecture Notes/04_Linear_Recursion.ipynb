# Linear Recursion

**Concept**: construct the solution to a problem by leveraging solutions of the same problem, but smaller in size.

**An example**: construct the solution for `find_max` using a solution of `find_max` with a smaller input.


```python
def find_max(L):
    if len(L)==1:
        return L[0]
    first = L[0]
    max_of_remaining_items = find_max(L[1:])
    if first > max_of_remaining_items:
        return first
    else:
        return max_of_remaining_items
```

```python
find_max([1,4,5,6,2,10,3])
```

```python
#PID:5
#
# Find the length of a list by adding 1 to the length of the remaining list
#
def length(L):
    if L==[]:
        return 0
    return 1 + length(L[1:])

length([1,20,39])
```

**Concept**: construct the solution to a problem by leveraging solutions of the same problem, but *smaller in size*.

We solve a problem using this technique by reducing the problem size.  Reducing problem size is critical because it implies progress. If we make a recursive with the same input (same size), the program won't work.

Reducing size implies that at some point, the input cannot be reduced further.  We have to solve these smallest cases (or "base cases") directly.

Naturally, we do not think about the smallest cases first.  We naturally think about how to solve a big problem using solutions of the same problems (but smaller).

But in terms of coding, we need to address/solve the smallest cases first.


```python
#PID:6
#
# Count the frequency of an item x recursively: comparing x to first element, and then
# using the same procedure to find the frequency of x in the remaining element
#
def count_freq(L, x):
    if L==[]:
        return 0
    if x == L[0]:
        return 1 + count_freq(L[1:],x)
    else:
        return count_freq(L[1:],x)

count_freq([1,2,3,2,10,2], 2) # output is 3
```

```python
#PID:7
#
# Analyze this example and use this analysis to design a recursive program to reverse a list
# L = [1,2,3,4,5].  Output we want is: [5,4,3,2,1]
# first item is 1
# What if you reverse the remaining item, we get [5,4,3,2].
# How do we get the correct output?  append 1 to [5,4,3,2]
# General strategy: append first to the reverse of the remaining list
#
def reverse(L):
    if L==[]:
        return []
    return reverse(L[1:]) + [L[0]]

reverse([1,2,3,4])
```

```python
#
# Find x in unsorted list
#
def find(x, L):
    if L==[]:
        return False
    if x==L[0]:
        return True
    return find(x, L[1:])
    
find(30, [0,20,10,5]) # output is True

# Find x in sorted list
find(10, [0,1,5,9,10,15,20,100,115]) # output is True

```

```python
#PID:8
#
# Find x in sorted list
# - compare x to the middle element (len(L)//2)
# - split list in halves right at the middle (left & right)
# - find x either in left or right
#
def find(x, L):
    if L==[]:
        return False
    
    mid = len(L)//2
    if x==L[mid]:
        return True
    if x < L[mid]:
        return find(x, L[0:mid])
    else:
        return find(x, L[mid+1:])

find(10, [0,1,5,9,10,15,20,100,115]) # output is True

```

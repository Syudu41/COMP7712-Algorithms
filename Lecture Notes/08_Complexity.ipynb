# Complexity of Running Time Equations

There are 3 measures of complexity of running time functions:
* Big O ($O$) -- estimate of upper bounds
* Omega ($\Omega$) -- estimate of lower bounds
* Theta ($\Theta$) -- estimate of tight bounds

Key thing to understand: these measures the growth of functions---when $n$ is large.

### Mathematical defintions

#### O -- upper bound
When we say that $T(n) \in O( f(n) )$, this means
* $f(n)$ is an upper bound (function) of $T(n)$.
* $T(n)$ is at most $f(n)$ times some constant
* More specifically, $T(n) \le c * f(n)$ for all large values of $n$.
    + $c$ has to be a specific number.
    + "for all large values of $n$" means e.g. all $n > 5$.


---
Example: prove that $T(n) = 2n^2 + 5 \in O(n^2)$

Before we attempt to prove this, we should understand that we have to prove $n^2$ is an upper bound of $T(n)$.

$T(n) = 2n^2 + 5 \le 2n^2 + 5n^2 = 7n^2$ for all $n > 1$.

We found $c$ to be 7.  $c=7$ is a certificate that shows $n^2$ to be an upper bound of $T(n)$.

Conclusion: $T(n) \in O(n^2)$

#### Definition of $O$

If $T(n) \in O(f(n))$, then there is a specific number $c$ such that $T(n) \le c * f(n)$ for all large values of $n$.

We actually have to find two numbers: (1) $c$ and (2) a threshold that implies "large values of $n$".

#### $\Theta$ -- tight bound

$T(n) \in \Theta(f(n))$ if $T(n) \in O(f(n))$ and $T(n) \in \Omega(f(n))$

This means $f(n)$ is both the upper bound and lower bound of $T(n)$.

Mathematically, there are two numbers $c_1$ and $c_2$ such that
$$c_1 * f(n) \le T(n) \le c_2 * f(n)$$ for all large values of $n$.

---

When we analyze the running of a complex program, the natural process is to identify loose bounds:
$$c_1 * f_1(n) \le T(n) \le c * f_2(n)$$ for all large values of $n$

If $f_1 \le f_2$, we don't a tight bound. We cannot determine $\Theta$

Let's upperbound $0.5n^2 + 15n$ with $n^2$.  Here's how:

$0.5n^2 + 15n \le 1*n^2 + 15*n^2 = 16n^2$

Let's lowerboud  $0.5n^2 + 15n$ with $n^2$.  Here's how:

$0.5n^2 + 15n \ge 0.4*n^2 + 0*n^2 = 0.4n^2$

Let's do this at the same time:
$0.4*n^2 \le 0.5*n^2 + 15 \le 16*n^2$ for all $n>1$

The lower bound constant is 0.4; upperbound constant is 16.

So $0.5*n^2 + 15 \in O(n^2)$ and $0.5*n^2 + 15 \in \Omega(n^2)$

Conclusion: $$0.5*n^2 + 15 \in \Theta(n^2)$$

# Non-linear Recursion

## What is nonlinear recursion?

In the previous lecture, we saw **linear recursion**: each function call makes **at most one** recursive call. This creates a chain of calls.

```
find_max([1,4,5,6,2])
  -> find_max([4,5,6,2])
      -> find_max([5,6,2])
          -> find_max([6,2])
              -> find_max([2])
```

**Nonlinear recursion** (also called **tree recursion** or **branching recursion**) occurs when a function makes **two or more** recursive calls per invocation. This creates a **tree** of calls, not a chain.

Why does this matter? Many natural structures are not linear — they branch. Trees, file systems, organizational charts, and decision processes all have this branching structure. To process them, we need to recurse into **each** branch.

The key idea remains the same: solve a problem by leveraging correct solutions to smaller subproblems. The only difference is that now we combine results from **multiple** subproblems instead of just one.

### Linear Recursion vs Iteration

````python
def find_max_rec(L):
    if len(L)==1:
        return L[0]
    first = L[0]
    max_of_remaining_list = find_max_rec(L[1:])
    if first > max_of_remaining_list:
        return first
    else:
        return max_of_remaining_list

def find_max_iter(L):
    m = L[0]
    for x in L:
        if x > m:
            m = x
    return m
````

````python
def binary_search_rec(x, L):
    if L==[]:
        return False
    mid = len(L)//2
    if x==L[mid]:
        return True
    if x < L[mid]:
        return binary_search_rec(x, L[0:mid])
    else:
        return binary_search_rec(x, L[mid+1:])
        
def binary_search_iter(x, L):
    if L==[]:
        return False
    left, right = 0, len(L)-1
    while right >= left:
        mid = (left+right)//2
        if x==L[mid]:
            return True
        if x < L[mid]:
            right = mid - 1
        else:
            left = mid + 1
    return False
    
binary_search_iter(7, [0,1,5,9,10,15,20,100,115]) # output is True

````

In general, recursive designs are easier to understand--you have acquire this new understanding of recurisve design first. But iterative designs are generally more efficient.

But for complex problems, non-linear recursive designs are much easier to understand.

## Binary Trees

A **binary tree** is a natural example of a branching structure. Each node has:
- A value (in our case, a color: "Blue" or "Red")
- A **left** subtree (which is itself a tree, or `None`)
- A **right** subtree (which is itself a tree, or `None`)

To process a binary tree, we must recurse into **both** the left and right subtrees. This is what makes it nonlinear.

Here is a class that generates a random binary tree:

````python
#PID:9
import random

#
# The parameter m controls how deep the tree is. 
# If m is not given, a random number is assigned to it.
#
class BTree:
    def __init__(self, m=None):
        if m is None:
            m = random.randint(2, 7)
        self.left = None
        self.right = None
        if random.randint(0, 1) == 0:
            self.color = 'Blue'
        else:
            self.color = 'Red'
        if random.randint(0, m) > 0:
            self.left = BTree(m-1)
        if random.randint(0, m) > 0:
            self.right = BTree(m-1)

    def print(self, indents=0):
        print('\t'*indents + self.color[0])
        if self.left is not None:
            self.left.print(indents+1)
        if self.right is not None:
            self.right.print(indents+1)

# print a random tree with 7 nodes
a_tree = BTree(3)
a_tree.print()
````

Notice the branching structure: each node can have a left child, a right child, both, or neither. A node with no children is called a **leaf**.

The `print` method itself is already an example of nonlinear recursion — it makes two recursive calls (one for `self.left`, one for `self.right`).

## The General Pattern

When processing a binary tree recursively, we almost always follow this pattern:

1. **Base case**: If the tree is `None`, return a base value.
2. **Recursive case**: Process the current node, recurse into the left subtree, recurse into the right subtree, and **combine** the results.

Compare this to linear recursion on lists:

| | Linear (Lists) | Nonlinear (Trees) |
|:---|:---|:---|
| **Structure** | `[first] + remaining` | current node + left subtree + right subtree |
| **Base case** | Empty list `[]` | Empty tree `None` |
| **Recursive calls** | One (on the remaining list) | Two (on left and right subtrees) |
| **Combine** | Combine first with one recursive result | Combine current node with two recursive results |

## Exercise 1: Count all nodes in a tree

Write a function `count_nodes(tree)` that returns the total number of nodes.

**Strategy in English**:
- If the tree is `None`, there are 0 nodes.
- Otherwise, the total number of nodes is: 1 (for the current node) + the number of nodes in the left subtree + the number of nodes in the right subtree.

Compare this to `length(L)` from the previous lecture:
- `length`: if `L` is empty, return 0. Otherwise, return 1 + length of the remaining list.
- `count_nodes`: if tree is `None`, return 0. Otherwise, return 1 + nodes in left + nodes in right.

The only difference is that we recurse into **two** subtrees instead of one remaining list.

````python
#PID:10

import random

class BTree:
    def __init__(self, m=None):
        if m is None:
            m = random.randint(2, 7)
        self.left = None
        self.right = None
        if random.randint(0, 1) == 0:
            self.color = 'Blue'
        else:
            self.color = 'Red'
        if random.randint(0, m) > 0:
            self.left = BTree(m-1)
        if random.randint(0, m) > 0:
            self.right = BTree(m-1)

    def print(self, indents=0):
        print('\t'*indents + self.color[0])
        if self.left is not None:
            self.left.print(indents+1)
        if self.right is not None:
            self.right.print(indents+1)
#
# What `count_nodes` does: it counts the number of nodes in a binary tree t
# (where t is the root node)
#
def count_nodes(t):
    if t is None:
        return 0
    return 1 + count_nodes(t.left) + count_nodes(t.right)

        
````

- `count_nodes(t)` counts the number of nodes of the tree rooted at `t`
- how does it work?  `count_nodes(t)` is equal 1 plus the number of nodes on the left subtree and the number of nodes of the right subtree.
- How do we calculate `the number of nodes on the left subtree` = `count_nodes(t.left)`

## Exercise 2: Height of a tree

Write a function `tree_height(tree)` that returns the height of a tree. The height is the longest path from the root to a leaf.

**Strategy in English**:
- If the tree is `None`, the height is 0.
- Otherwise, the height is 1 + ???

Why?

We are getting into this technique called **divide and conquer**.

We solve problem using solutions of subproblems.

- problem: `height` of a tree t
- a subproblem: `height` of the left subtree of t
- question: if we have the heights of the left/right subtrees of t, how do we calculate the height of t?
- `height(t)` = 1 + max(height of left, height of right)

````python
#PID:11
#
# returns the height of the tree (longest path from root to a leaf)
#
def height(tree):
    if tree is None:
        return -1
    height_of_left = height(tree.left)
    height_of_right = height(tree.right)
    return 1 + max(height_of_left, height_of_right)
````

````python
a_tree = BTree(3)
a_tree.print()
print("Height:", height(a_tree))
````

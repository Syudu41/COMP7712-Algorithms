### Jupyterbooks

These are text files in a json format.

A notebook is a list of cells.

Each cell is either text or code. Each cell has to be "executed" to see its output.

Text is written in Markdown syntax, which is a simplified version of HTML.

Code is written in Python syntax.

To use the notebooks, you have to know Markdown and Python syntax.

Shift+Return is a shortcut for executing a cell.


```python
print('this is a code cell.')
```

### Markdown syntax

Text cells are written in markdown syntax.

Texts are just regular texts.  But you might want to use Markdown syntax in some cases:
+ Creating lists of things (using + or - or *)
+ Highlighting **certain texts**
+ Making "headings" (HTML's h1, h2, h3, etc.):  use #, ##, ###

# this is level-1 heading

## this is level-2 heading

### this is level- heading

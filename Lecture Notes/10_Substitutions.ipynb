# Running Time of Recursive Algorithms

In Notebook 07, we analyzed the running time of **iterative** programs by counting steps inside loops.

For **recursive** programs, the running time equation is itself recursive. We call this a **recurrence relation**.

**Goal**: Given a recursive program, (1) write down its recurrence relation, and (2) solve it to find the complexity.

`````python
def count_occurrences(L, x):
    if L == []:                                    # constant
        return 0
    if L[0] == x:
        return 1 + count_occurrences(L[1:], x)     # ? 
    else:
        return count_occurrences(L[1:], x)         #
`````

Finding the running time of a program means: (1) finding the running time equation, and (2) estimate the complexity of that equation.

Running time equation $T(n) = a + ??$

$a$ is the constant that captures all constant running time.

To find the running time of `count_occurrences(L[1:], x)`, we have to know *what* T(n) describes.

**Definition**: $T(n)$ is running time of `count_occurrences(L, x)` where `n` is the input size (# of items in `L`).

What is the running time of `count_occurrences(L[1:], x)`? T(n-1)

$$T(n) = a + T(n-1)$$

`n` is a variable, it can have any value. We can replace `n` with `apple`: $T(apple) = a + T(apple - 1)$.

```
T(n) = a + T(n-1)
T(n) = a + a + T(n-2) = 2a + T(n-2)
T(n) = 2a + a + T(n-3) = 3a + T(n-3)
T(n) = 3a + a + T(n-4) = 4a + T(n-4)

After k steps: T(n) = ka + T(n-k)

After n steps: T(n) = n*a + T(n-n) = n*a + T(0) = n*a + b

T(0) is the running time of count_occurrences(L, x) when L has 0 items.  It's constant time: b.
```

The running time of `sum_list` with input size is $T(n)$.

# Divide and Conquer: From Brute Force to Efficient Solutions

In Notebook 11, we saw that the Master Theorem gives us a quick way to determine the complexity of divide-and-conquer algorithms:

$$T(n) = n^d + a \cdot T\left(\frac{n}{b}\right)$$

where:
- $a$ = number of recursive calls
- $b$ = factor by which input size shrinks
- $n^d$ = cost of non-recursive work (dividing + combining)

Today we will use this as a **diagnostic tool**. Many problems have $O(n^2)$ brute-force solutions. Can divide and conquer do better? The Master Theorem tells us exactly what determines the answer: the values of $a$, $b$, and $d$.

---

## Recap: The Master Theorem

Given $T(n) = n^d + a \cdot T\left(\frac{n}{b}\right)$, compare $d$ and $\log_b a$:

1. If $d \neq \log_b a$, then $T(n) \in \Theta\left(n^{\max(d,\; \log_b a)}\right)$

2. If $d = \log_b a$, then $T(n) \in \Theta(n^d \log n)$

Think of this as three "knobs" we can turn to improve efficiency:
- **Reduce $d$**: make the non-recursive work cheaper (e.g., a smarter combine step)
- **Reduce $a$**: make fewer recursive calls
- **Increase $b$**: shrink the input faster

## Problem 1: Sorting (Review)

We already know two ways to sort a list:

**Iterative approach** (selection sort, bubble sort): compare pairs of elements, swap as needed.
- Running time: $O(n^2)$

**Divide and conquer** (merge sort): split the list in half, sort each half recursively, merge the sorted halves.

```python
def merge_sort(L):
    if len(L) <= 1:
        return L
    mid = len(L) // 2
    sorted_left = merge_sort(L[:mid])
    sorted_right = merge_sort(L[mid:])
    return merge(sorted_left, sorted_right)   # merge costs O(n)
```

## Problem 4: Integer Multiplication

**Problem**: Multiply two $n$-digit numbers.

**Example**: Multiply 1234 × 5678. The grade-school algorithm multiplies each digit of one number by each digit of the other, then adds up the partial products. This takes $O(n^2)$ single-digit operations.

Can divide and conquer do better?

### D&C Decomposition

Split each $n$-digit number into two halves of $n/2$ digits.

For example, with $n = 4$:
- $1234 = 12 \cdot 10^2 + 34$
- $5678 = 56 \cdot 10^2 + 78$

In general, write:
- $x = a \cdot 10^{n/2} + b$
- $y = c \cdot 10^{n/2} + d$

where $a, b, c, d$ are $n/2$-digit numbers.

Then:
$$x \times y = (a \cdot 10^{n/2} + b)(c \cdot 10^{n/2} + d) = ac \cdot 10^n + (ad + bc) \cdot 10^{n/2} + bd$$

### Karatsuba's Trick (1960)

The bottleneck is $a = 4$. Can we compute $ac$, $ad + bc$, and $bd$ using **fewer** than 4 multiplications?

Recall we need three quantities: $ac$, $\;bd$, $\;$ and $\;ad + bc$.

**Key observation**:
$$(a + b)(c + d) = ac + ad + bc + bd$$

Therefore:
$$ad + bc = (a + b)(c + d) - ac - bd$$

So we can compute:
1. $p_1 = ac$ (one multiplication of $n/2$-digit numbers)
2. $p_2 = bd$ (one multiplication)
3. $p_3 = (a+b)(c+d)$ (one multiplication)
4. Then $ad + bc = p_3 - p_1 - p_2$ (only additions!)

**Three multiplications instead of four!**

## Problem 5: Closest Pair of Points

**Problem**: Given $n$ points in the plane, find the pair of points with the smallest distance between them.

**Example**: Given points $(1,1), (4,4), (2,2), (5,1), (3,5)$, the closest pair is $(1,1)$ and $(2,2)$ with distance $\sqrt{2}$.

### Brute Force Approach

Check every pair of points and compute the distance. There are $\binom{n}{2} = \frac{n(n-1)}{2}$ pairs.

##### Exercise 15

What is the running time of the brute-force approach?

### Divide and Conquer Approach

**Strategy**:
1. Sort the points by x-coordinate.
2. Split the points into a left half and a right half at the midpoint.
3. Recursively find the closest pair in each half.
4. **Combine**: Check if there is an even closer pair that crosses the dividing line.

The combine step is the challenge. Let $\delta$ = the smaller of the two distances found in steps 3.

If there is a crossing pair closer than $\delta$, both points must be within distance $\delta$ of the dividing line. We call this region the **strip**.

```
         |  strip  |
         |<--δ--|--δ-->|
         |         |
   Left  |    |    |  Right
         |    |    |
         |         |
         |  dividing line
```

### The Geometric Insight

It turns out we don't need to check all pairs in the strip!

**Key fact**: If we sort the strip points by y-coordinate, each point only needs to be compared to at most **7** other points (the next 7 points in the y-sorted order).

Why? Within a $\delta \times 2\delta$ rectangle around the dividing line, we can fit at most 8 points (since any two points on the same side are at least $\delta$ apart). So for each point, we only check a constant number of neighbors.

This means the strip check costs $O(n)$ (after sorting the strip by y), not $O(n^2)$!

## Summary: Improving Efficiency with Divide and Conquer

Every divide-and-conquer algorithm has the recurrence $T(n) = n^d + a \cdot T(n/b)$. The Master Theorem tells us the complexity depends on $a$, $b$, and $d$.

| Problem | Brute Force | D&C Approach | Bottleneck | Fix | Result |
|:---|:---|:---|:---|:---|:---|
| Sorting | $O(n^2)$ | $T = n + 2T(n/2)$ | none | — | $O(n \log n)$ |
| Max sublist sum | $O(n^2)$ | $T = n + 2T(n/2)$ | none | — | $O(n \log n)$ |
| Majority element | $O(n^2)$ | $T = n + 2T(n/2)$ | none | — | $O(n \log n)$ |
| Integer mult. | $O(n^2)$ | $T = n + 4T(n/2)$ | $a = 4$ | reduce $a$: $4 \to 3$ | $O(n^{1.585})$ |
| Closest pair | $O(n^2)$ | $T = n^2 + 2T(n/2)$ | $d = 2$ | reduce $d$: $2 \to 1$ | $O(n \log n)$ |

**The unified lesson**:
- For some problems (sorting, max sublist, majority), a straightforward D&C with a linear combine step already beats brute force.
- For other problems (integer multiplication, closest pair), a naive D&C is no better than brute force. The Master Theorem diagnoses the bottleneck: is $a$ too large, or is $d$ too large? Then we look for a clever fix.
